```python
import pandas as pd
import numpy as np

def analizar_reversion_vwap(file_path, sds_to_test=[1.0, 1.5, 2.0, 2.5, 3.0]):
    """
    Analiza la probabilidad de retorno al VWAP tras exceder bandas de SD.
    Optimizado para grandes volúmenes de datos.
    """
    try:
        print(f"📂 Cargando datos desde: {file_path}...")
        df = pd.read_parquet(file_path)
        
        # Estandarización de índice
        if not pd.api.types.is_datetime64_any_dtype(df.index):
            df.index = pd.to_datetime(df.index)
        
        # Filtro de sesión regular (09:30 - 16:00)
        df = df.between_time('09:30', '16:00').copy()
        df['date'] = df.index.date
        df['year'] = df.index.year

        print("⚡ Calculando métricas vectorizadas...")
        
        # Cálculo eficiente de VWAP y SD acumulada por grupo de fecha
        # Ponderación: (Precio * Volumen) y (Precio^2 * Volumen)
        df['pv'] = df['Close'] * df['Volume']
        df['pv2'] = (df['Close']**2) * df['Volume']
        
        grouped = df.groupby('date')
        
        df['cum_v'] = grouped['Volume'].cumsum()
        df['cum_pv'] = grouped['pv'].cumsum()
        df['cum_pv2'] = grouped['pv2'].cumsum()
        
        df['vwap'] = df['cum_pv'] / df['cum_v']
        # Fórmula de varianza: E[X^2] - (E[X])^2
        df['vwap_std'] = np.sqrt(np.maximum(0, (df['cum_pv2'] / df['cum_v']) - df['vwap']**2))

        results = []

        for x in sds_to_test:
            print(f"🔍 Testeando umbral: {x} SD")
            
            # Marcamos velas que exceden el umbral
            df['upper'] = df['vwap'] + (df['vwap_std'] * x)
            df['lower'] = df['vwap'] - (df['vwap_std'] * x)
            df['exceso'] = (df['High'] > df['upper']) | (df['Low'] < df['lower'])
            
            # Iteración por día para verificar reversión
            for (year, date), day_data in df.groupby(['year', 'date']):
                excesos = day_data[day_data['exceso']]
                
                if not excesos.empty:
                    # Hora del primer exceso del día
                    t_exceso = excesos.index[0]
                    # Datos desde ese momento hasta el cierre
                    ventana_reversion = day_data.loc[t_exceso:]
                    
                    # ¿El precio tocó el VWAP en la ventana restante?
                    exito = ((ventana_reversion['Low'] <= ventana_reversion['vwap']) & 
                             (ventana_reversion['High'] >= ventana_reversion['vwap'])).any()
                    
                    results.append({'Year': year, 'SD': x, 'Success': int(exito)})

        # Procesamiento de resultados
        final_df = pd.DataFrame(results)
        report = final_df.groupby(['Year', 'SD'])['Success'].mean() * 100
        
        return report.unstack()

    except Exception as e:
        print(f"🛑 Error en el análisis: {e}")
        return None

# --- PARÁMETROS DE EJECUCIÓN ---
FILE_PATH = '/home/quant/data/processed/nq_1m_continuous.parquet'
umbrales = [1.0, 1.5, 2.0, 2.5, 3.0]

tabla_resultados = analizar_reversion_vwap(FILE_PATH, umbrales)

if tabla_resultados is not None:
    print("\n📈 PROBABILIDAD DE RETORNO AL VWAP (%)")
    print(tabla_resultados.round(2).to_string())
```

```python
import pandas as pd
import numpy as np

def analizar_reversion_vwap(file_path, sds_to_test=[1.0, 1.5, 2.0, 2.5, 3.0]):
    """
    Analiza la probabilidad de retorno al VWAP tras haber salido del rango de SD.
    """
    try:
        print(f"📂 Cargando datos desde: {file_path}...")
        df = pd.read_parquet(file_path)
        
        # Estandarización de índice
        if not pd.api.types.is_datetime64_any_dtype(df.index):
            df.index = pd.to_datetime(df.index)
        
        # Filtro de sesión regular (09:30 - 16:00)
        df = df.between_time('09:30', '16:00').copy()
        df['date'] = df.index.date
        df['year'] = df.index.year

        print("⚡ Calculando métricas vectorizadas...")
        
        df['pv'] = df['Close'] * df['Volume']
        df['pv2'] = (df['Close']**2) * df['Volume']
        
        grouped = df.groupby('date')
        df['cum_v'] = grouped['Volume'].cumsum()
        df['cum_pv'] = grouped['pv'].cumsum()
        df['cum_pv2'] = grouped['pv2'].cumsum()
        
        df['vwap'] = df['cum_pv'] / df['cum_v']
        df['vwap_std'] = np.sqrt(np.maximum(0, (df['cum_pv2'] / df['cum_v']) - df['vwap']**2))

        all_results = []

        for x in sds_to_test:
            print(f"🔍 Testeando umbral: {x} SD")
            
            df['upper'] = df['vwap'] + (df['vwap_std'] * x)
            df['lower'] = df['vwap'] - (df['vwap_std'] * x)
            
            # El precio está fuera del rango (Extensión confirmada)
            df['fuera_rango'] = (df['High'] > df['upper']) | (df['Low'] < df['lower'])
            
            daily_stats = []
            
            for (year, date), day_data in df.groupby(['year', 'date']):
                # Buscamos el primer momento donde el precio ESTÁ FUERA del rango
                momentos_fuera = day_data.index[day_data['fuera_rango']]
                
                if len(momentos_fuera) > 0:
                    t_primera_salida = momentos_fuera[0]
                    
                    # Analizamos qué pasa DESPUÉS de esa primera salida
                    # Queremos ver si vuelve a tocar el VWAP antes del cierre
                    ventana_reversion = day_data.loc[t_primera_salida:].iloc[1:] 
                    
                    if not ventana_reversion.empty:
                        # Reversión exitosa: El High/Low cruza o toca la línea del VWAP
                        exito = ((ventana_reversion['Low'] <= ventana_reversion['vwap']) & 
                                 (ventana_reversion['High'] >= ventana_reversion['vwap'])).any()
                        
                        daily_stats.append({'Year': year, 'Success': int(exito)})
            
            if daily_stats:
                temp_df = pd.DataFrame(daily_stats)
                res = temp_df.groupby('Year')['Success'].mean() * 100
                res.name = x
                all_results.append(res)

        if all_results:
            reporte_final = pd.concat(all_results, axis=1)
            reporte_final.columns.name = 'SD'
            return reporte_final
        return None

    except Exception as e:
        print(f"🛑 Error en el análisis: {e}")
        return None

# --- PARÁMETROS DE EJECUCIÓN ---
FILE_PATH = '/home/quant/data/processed/nq_1m_continuous.parquet'
umbrales = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5] # Agregamos 3.5 por curiosidad estadística

tabla_resultados = analizar_reversion_vwap(FILE_PATH, umbrales)

if tabla_resultados is not None:
    print("\n📈 PROBABILIDAD DE RETORNO AL VWAP (%)")
    print("Condición: El precio debe salir del rango de SD y luego tocar el VWAP.")
    print(tabla_resultados.round(2).to_string())
```

```python
import pandas as pd
import numpy as np

def analizar_reversion_vwap(file_path, sds_to_test=[1.0, 1.5, 2.0, 2.5, 3.0], timeframe='1min'):
    """
    Analiza la probabilidad de retorno al VWAP tras haber salido del rango de SD.
    Utiliza el 'Typical Price' para aproximar la precisión de ticks usando velas.
    """
    try:
        print(f"📂 Cargando datos desde: {file_path}...")
        df_raw = pd.read_parquet(file_path)
        
        if not pd.api.types.is_datetime64_any_dtype(df_raw.index):
            df_raw.index = pd.to_datetime(df_raw.index)
        
        # --- PRECISIÓN: Cálculo de base sobre 1min ---
        # El Typical Price (H+L+C)/3 es la mejor aproximación a los datos por tick
        df_raw['tp'] = (df_raw['High'] + df_raw['Low'] + df_raw['Close']) / 3
        df_raw['pv'] = df_raw['tp'] * df_raw['Volume']
        df_raw['pv2'] = (df_raw['tp']**2) * df_raw['Volume']

        # Filtro de sesión regular NY (09:30 - 16:00)
        df_raw = df_raw.between_time('09:30', '16:00').copy()
        df_raw['date'] = df_raw.index.date
        
        # Cálculo acumulativo diario (Reinicio de VWAP cada mañana)
        print("🧮 Calculando VWAP y SD acumulados...")
        grouped_raw = df_raw.groupby('date')
        df_raw['cum_v'] = grouped_raw['Volume'].cumsum()
        df_raw['cum_pv'] = grouped_raw['pv'].cumsum()
        df_raw['cum_pv2'] = grouped_raw['pv2'].cumsum()
        
        df_raw['vwap'] = df_raw['cum_pv'] / df_raw['cum_v']
        # Varianza ponderada por volumen: E[X^2] - (E[X])^2
        df_raw['vwap_std'] = np.sqrt(np.maximum(0, (df_raw['cum_pv2'] / df_raw['cum_v']) - df_raw['vwap']**2))

        # --- RESAMPLING PARA EL ANÁLISIS ESTRATÉGICO ---
        if timeframe != '1min':
            print(f"🔄 Resampleando lógica a {timeframe} para filtrar ruido...")
            df = df_raw.resample(timeframe).agg({
                'Open': 'first',
                'High': 'max',
                'Low': 'min',
                'Close': 'last',
                'vwap': 'last',      
                'vwap_std': 'last'   
            }).dropna()
        else:
            df = df_raw.copy()

        df['year'] = df.index.year
        all_results = []

        # Barrido de umbrales de Desviación Estándar
        for x in sds_to_test:
            print(f"🔍 Analizando umbral: {x} SD")
            df['upper'] = df['vwap'] + (df['vwap_std'] * x)
            df['lower'] = df['vwap'] - (df['vwap_std'] * x)
            
            # Condición de EXCESO: Cierre de vela fuera de la banda
            df['exceso'] = (df['Close'] > df['upper']) | (df['Close'] < df['lower'])
            
            daily_stats = []
            for (year, date), day_data in df.groupby(['year', 'date']):
                # Primer momento del día con cierre en exceso
                excesos = day_data.index[day_data['exceso']]
                
                if len(excesos) > 0:
                    t_exceso = excesos[0]
                    # Monitoreo de reversión DESPUÉS de la vela de señal
                    ventana_reversion = day_data.loc[t_exceso:].iloc[1:] 
                    
                    if not ventana_reversion.empty:
                        # Éxito: El rango de precios de la vela toca la línea del VWAP
                        exito = ((ventana_reversion['Low'] <= ventana_reversion['vwap']) & 
                                 (ventana_reversion['High'] >= ventana_reversion['vwap'])).any()
                        daily_stats.append({'Year': year, 'Success': int(exito)})
            
            if daily_stats:
                temp_df = pd.DataFrame(daily_stats)
                res = temp_df.groupby('Year')['Success'].mean() * 100
                res.name = x
                all_results.append(res)

        # Generación del Reporte Final
        if all_results:
            reporte_final = pd.concat(all_results, axis=1)
            reporte_final.columns.name = 'SD'
            return reporte_final
        return None

    except Exception as e:
        print(f"🛑 Error crítico: {e}")
        return None

# --- EJECUCIÓN DEL SCRIPT ---
FILE_PATH = '/home/quant/data/processed/nq_1m_continuous.parquet'
TF_ANALISIS = '5min'  # Recomendado 5m o 15m para cuentas de fondeo
umbrales = [1.5, 2.0, 2.5, 3.0]

tabla_resultados = analizar_reversion_vwap(FILE_PATH, umbrales, timeframe=TF_ANALISIS)

if tabla_resultados is not None:
    print(f"\n📊 RESULTADOS: PROBABILIDAD DE REVERSIÓN AL VWAP (%)")
    print(f"Configuración: Datos base 1m | Señal en {TF_ANALISIS}")
    print("=" * 60)
    print(tabla_resultados.round(2).to_string())
    print("=" * 60)
```

```python
import pandas as pd
import numpy as np

def calcular_rsi(series, period=2):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def analizar_reversion_vwap_con_rsi(file_path, sds_to_test=[2.0], timeframe='5min', stops_to_test=[30, 50, 80, 100, 120]):
    """
    Analiza la reversión al VWAP comparando múltiples niveles de Stop Loss.
    Calcula Win Rate, MAE, PnL y Profit Factor para evaluar la calidad del riesgo.
    """
    try:
        print(f"📂 Cargando datos desde: {file_path}...")
        df_raw = pd.read_parquet(file_path)
        
        if not pd.api.types.is_datetime64_any_dtype(df_raw.index):
            df_raw.index = pd.to_datetime(df_raw.index)
        
        # --- CÁLCULO TÉCNICO (VWAP & Bandas) ---
        df_raw['tp'] = (df_raw['High'] + df_raw['Low'] + df_raw['Close']) / 3
        df_raw['pv'] = df_raw['tp'] * df_raw['Volume']
        df_raw['pv2'] = (df_raw['tp']**2) * df_raw['Volume']

        df_raw = df_raw.between_time('09:30', '16:00').copy()
        df_raw['date'] = df_raw.index.date
        
        grouped_raw = df_raw.groupby('date')
        df_raw['cum_v'] = grouped_raw['Volume'].cumsum()
        df_raw['cum_pv'] = grouped_raw['pv'].cumsum()
        df_raw['cum_pv2'] = grouped_raw['pv2'].cumsum()
        
        df_raw['vwap'] = df_raw['cum_pv'] / df_raw['cum_v']
        df_raw['vwap_std'] = np.sqrt(np.maximum(0, (df_raw['cum_pv2'] / df_raw['cum_v']) - df_raw['vwap']**2))

        # Resample a 5min para mejorar la ejecución y reducir el ruido
        if timeframe != '1min':
            df = df_raw.resample(timeframe).agg({
                'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last',
                'vwap': 'last', 'vwap_std': 'last'   
            }).dropna()
        else:
            df = df_raw.copy()

        # Filtro de años para el backtest
        df = df[df.index.year < 2025].copy()
        
        df['rsi2'] = calcular_rsi(df['Close'], 2)
        df['year'] = df.index.year
        df['date'] = df.index.date
        
        final_reports = {}

        for sl_pts in stops_to_test:
            all_results = []
            for x in sds_to_test:
                print(f"🔍 Ejecutando Simulación: {x} SD | Stop Loss: {sl_pts} pts")
                df['upper'] = df['vwap'] + (df['vwap_std'] * x)
                df['lower'] = df['vwap'] - (df['vwap_std'] * x)
                df['en_exceso_sup'] = df['Close'] > df['upper']
                df['en_exceso_inf'] = df['Close'] < df['lower']
                
                daily_stats = []
                
                for (year, date), day_data in df.groupby(['year', 'date']):
                    # Gatillo: Cruce de banda previo + Reentrada con RSI
                    trigger_long = day_data[(day_data['en_exceso_inf'].shift(1)) & (day_data['rsi2'] > 10)]
                    trigger_short = day_data[(day_data['en_exceso_sup'].shift(1)) & (day_data['rsi2'] < 90)]
                    
                    triggers = pd.concat([trigger_long, trigger_short]).sort_index()
                    
                    if not triggers.empty:
                        t_entrada = triggers.index[0]
                        precio_entrada = triggers.iloc[0]['Close']
                        es_long = t_entrada in trigger_long.index
                        ventana = day_data.loc[t_entrada:]
                        
                        resultado = 0 
                        pnl_puntos = 0
                        max_drawdown_trade = 0
                        
                        for idx, row in ventana.iterrows():
                            # Seguimiento del MAE (Maximum Adverse Excursion)
                            current_mae = precio_entrada - row['Low'] if es_long else row['High'] - precio_entrada
                            max_drawdown_trade = max(max_drawdown_trade, current_mae)
                            
                            # Lógica de Salida: Stop Loss
                            if max_drawdown_trade >= sl_pts:
                                resultado = -1
                                pnl_puntos = -sl_pts
                                break
                            
                            # Lógica de Salida: Take Profit (Regresión a la Media - VWAP)
                            target_hit = (es_long and row['High'] >= row['vwap']) or (not es_long and row['Low'] <= row['vwap'])
                            if target_hit:
                                resultado = 1
                                pnl_puntos = abs(precio_entrada - row['vwap'])
                                break
                        
                        daily_stats.append({
                            'Year': year, 
                            'Win': 1 if resultado == 1 else 0,
                            'Loss': 1 if resultado == -1 else 0,
                            'MAE_Exitosos': max_drawdown_trade if resultado == 1 else np.nan,
                            'PnL_Ganado': pnl_puntos if resultado == 1 else 0,
                            'PnL_Perdido': abs(pnl_puntos) if resultado == -1 else 0,
                            'PnL_Total': pnl_puntos
                        })
                
                if daily_stats:
                    temp_df = pd.DataFrame(daily_stats)
                    res = temp_df.groupby('Year').agg({
                        'Win': 'mean',
                        'MAE_Exitosos': lambda x: np.nanpercentile(x, 90) if not x.isnull().all() else 0,
                        'PnL_Total': 'sum',
                        'PnL_Ganado': 'sum',
                        'PnL_Perdido': 'sum'
                    })
                    
                    # Cálculo del Profit Factor (Beneficio Bruto / Pérdida Bruta)
                    res['ProfitFactor'] = res['PnL_Ganado'] / res['PnL_Perdido'].replace(0, np.nan)
                    
                    res = res[['Win', 'MAE_Exitosos', 'PnL_Total', 'ProfitFactor']]
                    res.columns = [f'WinRate_{x}', f'MAE_P90_{x}', f'PnL_{x}', f'PF_{x}']
                    res[f'WinRate_{x}'] *= 100
                    all_results.append(res)
            
            final_reports[sl_pts] = pd.concat(all_results, axis=1)

        return final_reports

    except Exception as e:
        print(f"🛑 Error crítico en el procesamiento: {e}")
        return None

# --- CONFIGURACIÓN DE BACKTEST ---
FILE_PATH = '/home/quant/data/processed/nq_1m_continuous.parquet'
STOPS_FINAL_TEST = [30, 50, 80, 100, 120]

resultados = analizar_reversion_vwap_con_rsi(
    file_path=FILE_PATH, 
    sds_to_test=[2.0], 
    timeframe='5min', 
    stops_to_test=STOPS_FINAL_TEST
)

if resultados:
    for sl, df_res in resultados.items():
        print(f"\n📊 MÉTRICAS CON STOP LOSS: {sl} PUNTOS")
        print("PF: Profit Factor | MAE_P90: Ruido máximo en el 90% de trades ganadores")
        print("-" * 100)
        print(df_res.round(2).to_string())
        print("-" * 100)
```

```python
import pandas as pd
import numpy as np
import os
from datetime import time

# --- CONFIGURACIÓN DE RUTAS ---
# Asegúrate de que esta ruta sea la correcta en tu entorno
PARQUET_PATH = '/home/quant/data/processed/nq_1m_continuous.parquet'

# --- PARÁMETROS DEL SISTEMA 966R (CAPA 2 Y 3) ---
HORA_LIMITE_ENTRADA = time(11, 0)
VENTANA_SENTINEL = 15
UMBRAL_STRIKE_RATE = 0.32

# --- CONFIGURACIÓN DEL MOTOR HÍBRIDO ---
# Valores de salida en unidades R
STOP_LOSS_RAPIDO = -0.45    # Salida por EMA7 o < 1R
TAKE_PROFIT_EXTREMO = 2.8   # Salida por EMA21 o > 2R
TARGET_PROMEDIO = 1.1      # Salida entre 1R y 2R

# Configuración de visualización de Pandas
pd.options.display.max_columns = None
pd.options.display.float_format = '{:.2f}'.format

print("✅ Celda 1: Entorno y Parámetros Globales configurados.")
```

```python
def preparar_entorno_profesional():
    print("🚀 Capa 0: Iniciando carga de datos y cálculo de indicadores...")
    
    if not os.path.exists(PARQUET_PATH):
        raise FileNotFoundError(f"❌ No se encontró el archivo Parquet en: {PARQUET_PATH}")
    
    df = pd.read_parquet(PARQUET_PATH)
    df.index = pd.to_datetime(df.index)
    
    # Limpieza de zona horaria para evitar conflictos en el join
    if df.index.tz is not None: 
        df.index = df.index.tz_localize(None)
    df = df.tz_localize('America/New_York', ambiguous='infer')

    # --- EMAs de 1 Minuto (Momentum) ---
    for p in [10, 20, 50, 100]:
        df[f'EMA1m_{p}'] = df['Close'].ewm(span=p, adjust=False).mean()

    # --- EMAs de 1 Hora (Tendencia Estructural) ---
    # Resampleamos a 1H, calculamos EMA y luego re-expandimos a 1m
    df_1h = df['Close'].resample('1h').last().to_frame()
    for p in [20, 50, 100, 200]:
        df_1h[f'EMA1h_{p}'] = df_1h['Close'].ewm(span=p, adjust=False).mean()
    
    # Unir datos de 1H al dataframe de 1m y propagar valores (ffill)
    df = df.join(df_1h.drop(columns='Close'), how='left')
    cols_1h = [f'EMA1h_{p}' for p in [20, 50, 100, 200]]
    df[cols_1h] = df[cols_1h].ffill()

    print(f"✅ Datos cargados: {len(df)} filas.")
    print(f"✅ Columnas técnicas listas: {[c for c in df.columns if 'EMA' in c]}")
    return df

# Ejecución
df_nq = preparar_entorno_profesional()
```

```python
def generar_master_df_limpio(df):
    print(f"🚀 Capa 1: Extrayendo trades del Midpoint...")
    print(f"⏱️ Analizando {df.index.normalize().unique().size} días de trading...")
    
    # Creamos una columna de fecha para agrupar eficientemente
    df_temp = df.copy()
    df_temp['date_only'] = df_temp.index.date
    grupos = df_temp.groupby('date_only')
    
    master_records = []
    # Columnas que capturaremos en el momento exacto de la entrada
    cols_filtros = [f'EMA1h_{p}' for p in [20, 50, 100, 200]] + \
                   [f'EMA1m_{p}' for p in [10, 20, 50, 100]]
    
    for fecha, day_data in grupos:
        if len(day_data) < 100: continue # Ignorar días incompletos
        
        # A. Definición de ORB (09:30 - 09:34)
        orb_data = day_data.between_time('09:30', '09:34')
        if orb_data.empty: continue
        
        oh, ol = orb_data['High'].max(), orb_data['Low'].min()
        om = (oh + ol) / 2
        r_unit = (oh - ol) / 2
        
        # VALIDACIÓN: Evitar división por cero si OH == OL
        if r_unit <= 0.0001: continue
        
        # B. Post-ORB (Búsqueda de confirmación y retest)
        post_orb = day_data.between_time('09:35', '16:00')
        if post_orb.empty: continue
        
        # Convertimos a matriz para velocidad (O:0, H:1, L:2, C:3)
        vals = post_orb[['Open', 'High', 'Low', 'Close'] + cols_filtros].values
        times_idx = post_orb.index.time
        
        confirmed, direction = False, None
        
        for i in range(len(vals)):
            # 1. Esperar confirmación de ruptura
            if not confirmed:
                if vals[i][3] > oh: confirmed, direction = True, 'Long'
                elif vals[i][3] < ol: confirmed, direction = True, 'Short'
                continue
            
            # 2. Buscar Retest al Midpoint (OM)
            is_retest = (direction == 'Long' and vals[i][2] <= om) or \
                        (direction == 'Short' and vals[i][1] >= om)
            
            if is_retest:
                # C. Cálculo de MFE (Máximo R alcanzado)
                resto_dia = vals[i:]
                if direction == 'Long':
                    max_p = resto_dia[:, 1].max()
                    mfe_r = (max_p - om) / r_unit
                else:
                    min_p = resto_dia[:, 2].min()
                    mfe_r = (om - min_p) / r_unit
                
                # D. Crear Registro de Trade
                res = {
                    'Date': fecha, 
                    'Entry_Time': times_idx[i], 
                    'Direction': direction,
                    'ORB_M': om, 
                    'R_Unit': r_unit, 
                    'Max_R_Alcanzado': mfe_r
                }
                
                # Inyectar valores de EMAs en el momento de la entrada
                for idx, col_name in enumerate(cols_filtros):
                    res[col_name] = vals[i][4 + idx]
                
                master_records.append(res)
                break # Solo un trade por día siguiendo la lógica 966R

    df_m = pd.DataFrame(master_records)
    print(f"✅ Master DF Finalizado: {len(df_m)} trades detectados.")
    return df_m

# Ejecución
df_master = generar_master_df_limpio(df_nq)
```

```python
def motor_966_final(df_m, filtro_tendencia='EMA1h_200'):
    print(f"🚀 Iniciando Motor 966R con filtro: {filtro_tendencia}")
    df = df_m.copy()
    
    # 1. Filtro Horario (Configurado en Celda 1)
    df = df[df['Entry_Time'] <= HORA_LIMITE_ENTRADA]
    
    # 2. Filtro de Tendencia (EMA 1H)
    mask = ((df['Direction'] == 'Long') & (df['ORB_M'] > df[filtro_tendencia])) | \
           ((df['Direction'] == 'Short') & (df['ORB_M'] < df[filtro_tendencia]))
    df = df[mask].copy()
    
    # 3. Sentinel W15 (Filtro de racha negativa)
    df['Win_T'] = (df['Max_R_Alcanzado'] >= 1.0).astype(int)
    df['SR_W15'] = df['Win_T'].rolling(window=VENTANA_SENTINEL).mean()
    df = df[df['SR_W15'] >= UMBRAL_STRIKE_RATE].copy()
    
    # 4. Lógica Híbrida 966R (Salidas asimétricas)
    def calcular_r_final(mfe):
        if mfe < 1.0: return STOP_LOSS_RAPIDO    # -0.45R
        if mfe >= 2.0: return TAKE_PROFIT_EXTREMO # 2.8R
        return TARGET_PROMEDIO                   # 1.1R
    
    df['R_Final'] = df['Max_R_Alcanzado'].apply(calcular_r_final)
    
    # 5. Métricas de Performance
    df['Year'] = pd.to_datetime(df['Date']).dt.year
    net_r = df['R_Final'].sum()
    curva = df['R_Final'].cumsum()
    max_dd = (curva.cummax() - curva).max()
    sharpe = (df['R_Final'].mean() / df['R_Final'].std()) * np.sqrt(252)
    
    print("\n" + "="*40)
    print(f"📊 REPORTE DE ESTRATEGIA: 966R")
    print("="*40)
    print(f"Trades Totales:  {len(df)}")
    print(f"Profit Neto:     {net_r:.2f} R")
    print(f"Max Drawdown:    {max_dd:.2f} R")
    print(f"Sharpe Ratio:    {sharpe:.2f}")
    print(f"R/Trade Promed:  {df['R_Final'].mean():.2f}")
    print("="*40)
    
    # Reporte Anual
    anual = df.groupby('Year')['R_Final'].agg(['count', 'sum']).rename(columns={'count':'Trades', 'sum':'R_Anual'})
    print(anual)
    
    return df

# Ejecución del sistema final
df_final = motor_966_final(df_master)
```

```python
import matplotlib.pyplot as plt

def generar_reporte_pro(df_f):
    print("📊 AUDITORÍA ESTADÍSTICA COMPLETA - SISTEMA 966R")
    print("-" * 50)
    
    # 1. Cálculos de Métricas
    total_trades = len(df_f)
    wins = df_f[df_f['R_Final'] > 0]
    losses = df_f[df_f['R_Final'] <= 0]
    
    win_rate = (len(wins) / total_trades) * 100
    avg_p = wins['R_Final'].mean()
    avg_l = abs(losses['R_Final'].mean())
    ratio_w_l = avg_p / avg_l if avg_l != 0 else 0
    
    net_r = df_f['R_Final'].sum()
    curva = df_f['R_Final'].cumsum()
    max_hasta_ahora = curva.cummax()
    drawdown_serie = curva - max_hasta_ahora # Valores negativos
    max_dd = drawdown_serie.min()
    
    # Sharpe Anualizado Realista (basado en trades)
    sharpe = (df_f['R_Final'].mean() / df_f['R_Final'].std()) * np.sqrt(252)

    # 2. Impresión de Resultados
    stats = {
        "Net Profit (R)": net_r,
        "Win Rate (%)": win_rate,
        "Total Trades": total_trades,
        "Avg Profit (R)": avg_p,
        "Avg Loss (R)": avg_l,
        "Ratio W/L": ratio_w_l,
        "Max Drawdown (R)": max_dd,
        "Recovery Factor": net_r / abs(max_dd) if max_dd != 0 else 0,
        "Sharpe Ratio": sharpe
    }
    
    for k, v in stats.items():
        print(f"{k:<20}: {v:>10.2f}")

    # 3. Visualización
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True, 
                                   gridspec_kw={'height_ratios': [3, 1]})

    # Gráfico 1: Curva de Capital
    ax1.plot(curva.values, label='R Acumulado', color='#2ecc71', lw=2)
    ax1.fill_between(range(len(curva)), curva.values, color='#2ecc71', alpha=0.1)
    ax1.set_title('Curva de Crecimiento (Unidades R)', fontsize=14)
    ax1.grid(alpha=0.3)
    ax1.legend()

    # Gráfico 2: Drawdown Underwater
    ax2.fill_between(range(len(drawdown_serie)), drawdown_serie.values, 0, 
                     color='#e74c3c', alpha=0.5)
    ax2.set_title('Drawdown Histórico (Underwater Chart)', fontsize=12)
    ax2.set_ylabel('R Caídos')
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

# Ejecutar reporte
generar_reporte_pro(df_final)
```

```python
def generar_analisis_temporal(df_f):
    df = df_f.copy()
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    
    # 1. Reporte Anual Completo
    print("📅 REPORTE DE CONSISTENCIA ANUAL")
    print("-" * 60)
    anual = df.groupby('Year').agg(
        Trades=('R_Final', 'count'),
        Net_R=('R_Final', 'sum'),
        Max_MFE=('Max_R_Alcanzado', 'max'),
        Win_Rate=('R_Final', lambda x: (x > 0).mean() * 100)
    )
    anual['R_Acumulado'] = anual['Net_R'].cumsum()
    print(anual)
    print("\n")

    # 2. Heatmap Mensual (Últimos 2 años: 2024 - 2025)
    # Nota: Ajusta los años si tu dataset cubre otros periodos
    años_interes = [2024, 2025]
    df_reciente = df[df['Year'].isin(años_interes)]
    
    if not df_reciente.empty:
        pivot_mensual = df_reciente.pivot_table(
            index='Year', 
            columns='Month', 
            values='R_Final', 
            aggfunc='sum'
        ).fillna(0)
        
        # Mapear nombres de meses
        meses_nombres = {
            1:'Ene', 2:'Feb', 3:'Mar', 4:'Abr', 5:'May', 6:'Jun',
            7:'Jul', 8:'Ago', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dic'
        }
        pivot_mensual.columns = [meses_nombres[m] for m in pivot_mensual.columns]
        
        print(f"📊 RENDIMIENTO MENSUAL (R) - ÚLTIMOS {len(años_interes)} AÑOS")
        print("-" * 60)
        # Usamos style para dar formato de tabla financiera
        display(pivot_mensual.style.background_gradient(cmap='RdYlGn', axis=None).format("{:.1f}"))
    else:
        print("⚠️ No hay suficientes datos para los años 2024-2025")

# Ejecución
generar_analisis_temporal(df_final)
```

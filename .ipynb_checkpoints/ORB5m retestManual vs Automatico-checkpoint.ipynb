```python
import pandas as pd
import numpy as np
from datetime import time, date

def calculate_orb_1m_v7_clean(day_data):
    """Calcula los niveles del Opening Range Breakout (ORB) de 9:30 a 9:34."""
    # Nota: Los tiempos deben ser ajustados si el ORB no es exactamente 9:30-9:34.
    orb_start_time = pd.to_datetime('09:30:00').time()
    orb_end_time = pd.to_datetime('09:34:00').time()
    
    orb_data = day_data[(day_data.index.time >= orb_start_time) & (day_data.index.time <= orb_end_time)].copy()
    
    if orb_data.empty or len(orb_data) < 5:
        # No hay suficientes velas entre 9:30:00 y 9:35:00
        return None, None, None, None, True 

    # Se calcula el High y Low de las 6 velas (9:30 a 9:35)
    ORB_H = orb_data['High'].max()
    ORB_L = orb_data['Low'].min()
    R_T = ORB_H - ORB_L
    ORB_M = ORB_L + (R_T / 2)

    return ORB_H, ORB_L, ORB_M, R_T, False

print("Función 'calculate_orb_1m_v7_clean' definida.")
```

```python
import pandas as pd
import os
from datetime import time

# --- Mapeo de R y Constantes (Asegúrate de que estas constantes estén definidas en tu entorno) ---
# --- Mapeo de R y Constantes ---
PL_TO_R = {
    'Profit': 2.0, 'Loss': -1.0, 'Break Even': 0.0, 'SL': -1.0, 
    'TP': 2.0, 'BE': 0.0, 'Profit ': 2.0, 'Loss ': -1.0, 
    'Break Even ': 0.0, 'No Trade': 0.0, 'No Trade ': 0.0
}
COL_DATE = 'FECHA'
COL_PL = 'P/L'
BE_R_MULTIPLE = 1.0

# --- Función de Carga (v2 - Modificada para devolver 'Exit_MANUAL') ---
def load_and_clean_manual_backtest_data_v2(file_path):
    # Manejo de rutas
    if not os.path.exists(file_path):
        alternate_path = file_path.replace(' (1).csv', '.csv')
        if os.path.exists(alternate_path):
            file_path = alternate_path
        else:
            print(f"❌ ERROR CRÍTICO: Archivo no encontrado en las rutas: {file_path} y {alternate_path}")
            return None
    
    try:
        manual_df_raw = pd.read_csv(file_path)
        manual_df_raw.columns = [col.strip() for col in manual_df_raw.columns] 
        manual_df_raw[COL_PL] = manual_df_raw[COL_PL].astype(str).str.strip()
        manual_df_raw['Manual_R'] = manual_df_raw[COL_PL].map(PL_TO_R).fillna(999) 
        manual_df_clean = manual_df_raw[manual_df_raw['Manual_R'] != 999].copy()
        
        # --- CAMBIO CRÍTICO: RENOMBRAR A 'Exit_MANUAL' ---
        manual_df_clean['Exit_MANUAL'] = manual_df_clean[COL_PL].apply(
            lambda x: 'NO TRADE' if 'No Trade' in x else x
        )
        # ----------------------------------------------------

        manual_df_clean['Date_Parsed'] = pd.to_datetime(
            manual_df_clean[COL_DATE], 
            format='%m/%d/%Y', 
            errors='coerce'
        )
        manual_df_clean = manual_df_clean.dropna(subset=['Date_Parsed']).copy()
        manual_df_clean['Date'] = manual_df_clean['Date_Parsed'].dt.strftime('%Y-%m-%d')
        
        # --- CAMBIO CRÍTICO: Devolver 'Exit_MANUAL' ---
        manual_df = manual_df_clean[['Date', 'Manual_R', 'Exit_MANUAL']].copy()
        print(f"✅ Archivo manual cargado exitosamente desde {os.path.basename(file_path)}. {len(manual_df)} días encontrados.")
        return manual_df

    except Exception as e:
        print(f"❌ ERROR durante la carga o procesamiento del CSV: {e}")
        return None

print("Bloque 1/3: Cargado Bakctesting Manual (Versión Corregida)")
```

```python
import pandas as pd
import numpy as np
from datetime import time

# --- Constantes V33 ---
BE_R_MULTIPLE = 1.0 # El BE se activa si el precio ha avanzado 1.0 R a favor (por toque).

print("✅ Constantes definidas (BE_R_MULTIPLE = 1.0).")

# =====================================================================
# 1. FUNCIÓN DE EJECUCIÓN OHLC/OLHC (V33 - Lógica de la Vela)
# =====================================================================

def check_sequential_ohlc_olhc_v33(row, direction, entry_price, R, SL_level, TP_level, be_triggered):
    # Lógica que chequea una sola barra (vela)
    current_sl = entry_price if be_triggered else SL_level
    target_r = 2.0
    risk_r = 1.0

    def hit_tp(): return {'Exit': 'TP', 'Retorno_R': target_r}
    def hit_stop(is_be): 
        exit_type = 'BE' if is_be else 'SL'
        exit_r = 0.0 if is_be else -risk_r
        return {'Exit': exit_type, 'Retorno_R': exit_r}

    is_tp_hit_in_bar = False
    is_sl_or_be_hit_in_bar = False
    is_be_activation_touch_in_bar = False
    
    # ... (Cálculos de toque de niveles idénticos a V33) ...
    if direction == 'Long':
        BE_1R_ACTIVATION = entry_price + (R * BE_R_MULTIPLE)
        is_tp_hit_in_bar = row['High'] >= TP_level
        if row['Low'] <= current_sl: is_sl_or_be_hit_in_bar = True
        is_be_activation_touch_in_bar = row['High'] >= BE_1R_ACTIVATION
    else: # Short
        BE_1R_ACTIVATION = entry_price - (R * BE_R_MULTIPLE)
        is_tp_hit_in_bar = row['Low'] <= TP_level
        if row['High'] >= current_sl: is_sl_or_be_hit_in_bar = True
        is_be_activation_touch_in_bar = row['Low'] <= BE_1R_ACTIVATION

    # ... (Lógica de toque dual y toque BE/SL idéntica a V33) ...
    if is_tp_hit_in_bar and is_sl_or_be_hit_in_bar:
        if direction == 'Long':
            if (row['High'] - row['Open']) >= (row['Open'] - row['Low']): return hit_tp(), be_triggered
            else: return hit_stop(be_triggered), be_triggered 
        else: # Short
            if (row['Open'] - row['Low']) >= (row['High'] - row['Open']): return hit_tp(), be_triggered
            else: return hit_stop(be_triggered), be_triggered

    if not be_triggered and is_be_activation_touch_in_bar and is_sl_or_be_hit_in_bar:
        if direction == 'Short':
            if (row['Open'] - row['Low']) >= (row['High'] - row['Open']): return None, True 
            else: return hit_stop(False), False
        elif direction == 'Long':
            if (row['High'] - row['Open']) >= (row['Open'] - row['Low']): return None, True
            else: return hit_stop(False), False
                
    if is_tp_hit_in_bar: return hit_tp(), be_triggered
    if not be_triggered and is_be_activation_touch_in_bar: be_triggered = True
    if is_sl_or_be_hit_in_bar: return hit_stop(be_triggered), be_triggered
            
    return None, be_triggered

print("✅ Función de Lógica de Vela (OHLC/OLHC V33) definida.")
```

```python
# =====================================================================
# 2. FUNCIÓN DE BACKTEST V33 (Wrapper Diario)
# =====================================================================

def run_orb_retest_v33(day_data, ORB_H, ORB_L, ORB_M, R_T):
    R = R_T / 2; target_r = 2.0
    TP_L = ORB_M + (R * target_r); SL_L = ORB_L
    TP_S = ORB_M - (R * target_r); SL_S = ORB_H
    
    trade_data = day_data[day_data.index.time > pd.to_datetime('09:35:00').time()].copy()
    if trade_data.empty: return None

    entry_price = ORB_M
    trade_start_index = None; breakout_triggered = False; direction = None
    
    # Búsqueda de Breakout y Retest
    for idx, row in trade_data.iterrows():
        if not breakout_triggered:
            if row['High'] >= ORB_H: breakout_triggered = True; direction = 'Long'
            elif row['Low'] <= ORB_L: breakout_triggered = True; direction = 'Short'
            
        if breakout_triggered and trade_start_index is None:
            if direction == 'Long' and row['Low'] <= ORB_M: trade_start_index = idx; break  
            elif direction == 'Short' and row['High'] >= ORB_M: trade_start_index = idx; break  
    
    # Manejar "NO TRADE" (No Retest o No Breakout)
    if trade_start_index is None: 
        # print(f"DEBUG: {day_data.index.normalize().date[0].strftime('%Y-%m-%d')} - NO TRADE (No Retest)")
        return {'Date': day_data.index.normalize().date[0].strftime('%Y-%m-%d'), 'Direction': direction if direction else 'N/A', 'Exit_Time': None, 'Exit': 'NO TRADE', 'Retorno_R': 0.0}

    # 2.2. EJECUCIÓN DEL TRADE a partir del Retest
    execution_data = trade_data[trade_data.index >= trade_start_index].copy()
    be_triggered = False
    
    if direction == 'Long': SL_level, TP_level = SL_L, TP_L
    else: SL_level, TP_level = SL_S, TP_S

    for idx, row in execution_data.iterrows():
        exit_result, new_be_triggered = check_sequential_ohlc_olhc_v33(
            row, direction, entry_price, R, SL_level, TP_level, be_triggered
        )
        be_triggered = new_be_triggered 

        if exit_result:
            # print(f"DEBUG: {day_data.index.normalize().date[0].strftime('%Y-%m-%d')} - Salida: {exit_result['Exit']}")
            return {
                'Date': day_data.index.normalize().date[0].strftime('%Y-%m-%d'),
                'Direction': direction,
                'Exit_Time': idx.strftime('%H:%M:%S'),
                'Exit': exit_result['Exit'],
                'Retorno_R': exit_result['Retorno_R']
            }
    
    # Salida por EOD (End of Day)
    return {'Date': day_data.index.normalize().date[0].strftime('%Y-%m-%d'), 'Direction': direction, 'Exit_Time': None, 'Exit': 'EOD', 'Retorno_R': 0.0}

print("✅ Función de Backtest Diario (run_orb_retest_v33) definida.")
```

```python
# Importaciones necesarias para el reporte (asumiendo que no están en el kernel)
import pandas as pd
import numpy as np
import os # Necesario si la función de carga manual lo usa

# =====================================================================
# 3. FUNCIÓN DE COMPARACIÓN V33 (Reporte Limpio)
# =====================================================================

def run_monthly_orb_retest_v33_clean_report(df_nq_1m, start_date_str, end_date_str, file_path_manual):
    all_results = []
    
    # 1. Preparación de fechas
    start_date = pd.to_datetime(start_date_str).tz_localize('America/New_York').normalize()
    end_date = pd.to_datetime(end_date_str).tz_localize('America/New_York').normalize()
    dates = pd.date_range(start=start_date, end=end_date, freq='B', tz='America/New_York')
    
    print(f"DEBUG: Fechas de análisis listas. Total de días hábiles: {len(dates)}")
    
    # 2. Ejecutar Backtest (V33)
    for current_date in dates:
        day_data_nq_1m = df_nq_1m[df_nq_1m.index.normalize() == current_date].copy()
        
        if day_data_nq_1m.empty: continue
        
        try:
            # DEPENDENCIA EXTERNA 1: calculate_orb_1m_v7_clean
            ORB_H, ORB_L, ORB_M, R_T, no_orb = calculate_orb_1m_v7_clean(day_data_nq_1m) 
        except NameError: 
            print("ERROR FATAL: calculate_orb_1m_v7_clean NO está definida. Abortando backtest.")
            return

        # Si no hay ORB, asumimos que run_orb_retest_v33 devolverá 'NO TRADE'.
        if no_orb or R_T == 0:
            result = {'Date': current_date.strftime('%Y-%m-%d'), 'Direction': 'N/A', 'Exit_Time': None, 'Exit': 'NO TRADE', 'Retorno_R': 0.0}
        else:
            result = run_orb_retest_v33(day_data_nq_1m, ORB_H, ORB_L, ORB_M, R_T)
            
        if result: all_results.append(result)
            
    results_df_v33 = pd.DataFrame(all_results)
    if results_df_v33.empty: 
        print("ERROR CRÍTICO: El DataFrame de resultados del código está vacío. El bucle de ejecución falló.")
        return

    print(f"DEBUG: Backtest diario completado. Resultados: {len(results_df_v33)} días procesados.")

    # 3. Cargar Datos Manuales
    try:
        # DEPENDENCIA EXTERNA 2: load_and_clean_manual_backtest_data_v2
        manual_df = load_and_clean_manual_backtest_data_v2(file_path_manual) # Llama a la función definida por separado
    except NameError:
        print("ERROR FATAL: load_and_clean_manual_backtest_data_v2 NO está definida. Abortando reporte.")
        return
        
    if manual_df is None: return

    # 4. Merge y Comparación
    merged_df = pd.merge(results_df_v33, manual_df, on='Date', how='inner', suffixes=('_CODE', '_MANUAL'))
    
    # Limpieza de etiquetas para comparación
    merged_df['Exit_MANUAL_Clean'] = merged_df['Exit_MANUAL'].apply(
        lambda x: 'BE' if 'Break Even' in str(x) else ('SL' if 'Loss' in str(x) else str(x).replace(' ', ''))
    )
    merged_df['Exit_CODE_Clean'] = merged_df['Exit'].replace({'EOD': 'NO TRADE'})
    merged_df['Exit_MANUAL_Clean'] = merged_df['Exit_MANUAL_Clean'].replace({'NO TRADE': 'NO TRADE'})

    merged_df['Exit_Match'] = merged_df['Exit_CODE_Clean'] == merged_df['Exit_MANUAL_Clean']
    merged_df['R_Match'] = merged_df['Retorno_R'].round(2) == merged_df['Manual_R'].round(2)
    merged_df['Total_Match'] = merged_df['Exit_Match'] & merged_df['R_Match']
    
    # 5. Impresión de Resultados
    total_discrepancias = (~merged_df['Total_Match']).sum()
    total_R_codigo = merged_df[merged_df['Exit_CODE_Clean'] != 'NO TRADE']['Retorno_R'].sum()
    total_R_manual = merged_df[merged_df['Exit_MANUAL_Clean'] != 'NO TRADE']['Manual_R'].sum()
    total_dias_analizados_con_trade = len(merged_df[merged_df['Exit_CODE_Clean'] != 'NO TRADE'])
    
    # LA LISTA DE COLUMNAS CORRECTA
    comparison_cols = ['Date', 'Direction', 'Exit_CODE_Clean', 'Retorno_R', 'Exit_MANUAL_Clean', 'Manual_R', 'Total_Match']
    
    print("\n\n=======================================================================")
    print(f"🎉 REPORTE FINAL DE DISCREPANCIAS (V33 - Limpio)")
    print("=======================================================================")
    print(f"Total de Días Analizados (Con Trade en Código): {total_dias_analizados_con_trade}")
    print(f"Discrepancias Totales (R y Exit): {total_discrepancias}")
    print("-------------------------------------------------------")
    print(f"Ganancia Total CÓDIGO V33 (en R): {total_R_codigo:.2f}R")
    print(f"Ganancia Total MANUAL (en R): {total_R_manual:.2f}R")
    print("=======================================================")

    print("\n--- Comparación Detallada Octubre y Noviembre (2025) ---")
    print(merged_df[comparison_cols].to_string(index=False, float_format=lambda x: f'{x:.2f}'))
    print("=======================================================")

    if total_discrepancias > 0:
        print(f"\n🚨 DÍAS RESTANTES CON DISCREPANCIAS EN RETORNO Y/O EXIT:")
        discrepancias = merged_df[~merged_df['Total_Match']]
        print(discrepancias[comparison_cols].to_string(index=False, float_format=lambda x: f'{x:.2f}'))


# --- Ejecución Principal ---
START_DATE = '2025-10-01'
END_DATE = '2025-11-30'
FILE_PATH = '/home/quant/Downloads/5m_ORB_retest_Backtest - Sheet3 (1).csv'

print("\n--- INICIANDO EJECUCIÓN DEL REPORTE FINAL ---")
try:
    # Asegúrate de que df_nq_1m esté cargado antes de llamar a esta función
    run_monthly_orb_retest_v33_clean_report(df_nq_1m, START_DATE, END_DATE, FILE_PATH)
except NameError as e:
    print(f"\n❌ ERROR DE ENTORNO CRÍTICO: df_nq_1m (Datos de Precio) NO está definida. Asegúrate de cargarlos. Error: {e}")
except Exception as e:
    print(f"\n❌ ERROR DESCONOCIDO DURANTE LA EJECUCIÓN PRINCIPAL: Ocurrió una excepción: {e}")
```

```python
import pandas as pd

def simulate_orb_trade_detail(fecha_analisis, df_nq_1m, check_func, orb_calc_func):
    
    print(f"==================================================")
    print(f"| INICIANDO ANÁLISIS DETALLADO: {fecha_analisis} |")
    print(f"==================================================")

    try:
        df_day = df_nq_1m.loc[fecha_analisis].copy()
        orb_levels_tuple = orb_calc_func(df_day)
    except KeyError:
        print(f"❌ Error: La fecha {fecha_analisis} no se encuentra o las funciones no están disponibles.")
        return

    # Desempaquetar tupla de 5 elementos (H, L, M, RT, Valid)
    try:
        orb_h, orb_l, orb_m, range_total, _ = orb_levels_tuple
        r_unit = range_total / 2 # Calcular Riesgo Unitario
    except ValueError:
        print("❌ Error de formato: La función calculate_orb_1m_v7_clean no devolvió 5 valores (H, L, M, R_T, Valid).")
        print(f"   Valores devueltos: {orb_levels_tuple}")
        return

    # 1. Detección de Breakout (09:36 a 10:00)
    df_breakout_check = df_day.between_time('09:36:00', '10:00:00')
    
    breakout_info = {'time': None, 'direction': None}
    
    # Encontrar el primer breakout (Long o Short)
    for index, row in df_breakout_check.iterrows():
        if row['High'] >= orb_h:
            if breakout_info['direction'] is None:
                breakout_info = {'time': index, 'direction': 'Long'}
            break
        if row['Low'] <= orb_l:
            if breakout_info['direction'] is None:
                breakout_info = {'time': index, 'direction': 'Short'}
            break
    
    # 2. Asignar Trade (usamos la dirección Short que tu código detectó para seguir depurando)
    if breakout_info['direction'] == 'Short':
        direction = 'Short'
        entry_price = orb_m
        sl_price = orb_h
        tp_price = entry_price - (2 * r_unit)
        be_activation_level = entry_price - r_unit
        retest_level = entry_price
        
        # Buscar Retest Short (High toca Entry Price)
        df_retest_check = df_day.loc[breakout_info['time']:].between_time('09:36:00', '10:00:00')
        retest_touch = df_retest_check[df_retest_check['High'] >= retest_level].head(1)
        
    elif breakout_info['direction'] == 'Long':
        direction = 'Long'
        entry_price = orb_m
        sl_price = orb_l
        tp_price = entry_price + (2 * r_unit)
        be_activation_level = entry_price + r_unit
        retest_level = entry_price
        
        # Buscar Retest Long (Low toca Entry Price)
        df_retest_check = df_day.loc[breakout_info['time']:].between_time('09:36:00', '10:00:00')
        retest_touch = df_retest_check[df_retest_check['Low'] <= retest_level].head(1)
    else:
        print("⛔ No se encontró un Breakout válido entre 09:36 y 10:00. Fin del análisis.")
        return

    print("\n--- 1. NIVELES Y DIRECCIÓN ---")
    print(f"   BREAKOUT DETECTADO: {breakout_info['direction']} en {breakout_info['time'].strftime('%H:%M:%S')}")
    print(f"   Dirección del Trade: {direction}")
    print(f"   Entrada (ORB Mid):  {entry_price:.2f}")
    print(f"   SL Inicial (1R):    {sl_price:.2f}")
    print(f"   BE Activación (1R): {be_activation_level:.2f}")
    print(f"   TP (2R):            {tp_price:.2f}")
    
    if retest_touch.empty:
        print(f"⛔ No hubo retest (entrada) después del breakout {direction} antes de las 10:00. Fin del análisis.")
        return
    
    entry_time = retest_touch.index[0]
    print(f"\n   ✅ Entrada (Retest) detectada en: {entry_time.strftime('%H:%M:%S')}")


    # 3. Simular la Ejecución Vela a Vela
    df_trade_execution = df_day.loc[entry_time:].iloc[1:]

    print("\n--- 2. SEGUIMIENTO VELA A VELA ---")
    
    sl_price_actual = sl_price
    be_triggered = False
    
    # Simulación del bucle interno de run_orb_retest_v33
    for index, row in df_trade_execution.iterrows():
        
        # Ejecutar la lógica de chequeo de secuencialidad V33 (LA CLAVE)
        exit_status, be_state_change = check_func(
            row, 
            direction, 
            sl_price_actual,
            tp_price, 
            be_activation_level, 
            entry_price, 
            be_triggered
        )
        
        # Imprimir estado de la vela ANTES DE LA SALIDA
        print(f"|--- {index.strftime('%H:%M:%S')} | O:{row['Open']:.2f} H:{row['High']:.2f} L:{row['Low']:.2f} C:{row['Close']:.2f} | SL Act: {sl_price_actual:.2f} | BE T:{be_triggered}")

        if exit_status is not None:
            # Trade cerrado (TP, SL o BE) 
            exit_name = exit_status['Exit']
            exit_r = exit_status['Retorno_R']
            
            # 🛑 ERROR DE LÓGICA V33: Si el código detecta TP en una vela que NO tocó niveles.
            # Imprimimos el resultado y detenemos la ejecución para la depuración.
            print(f"|--- 🛑 SALIDA: {exit_name} ({exit_r:.2f}R).")
            return
            
        elif be_state_change:
            # BE activado, movemos SL
            be_triggered = True
            sl_price_actual = entry_price # SL se mueve a la entrada (BE)
            print(f"|--- ⚠️ BE ACTIVADO. SL movido a la Entrada ({sl_price_actual:.2f}).")
            
    
    print("\n--- 3. RESULTADO FINAL DEL CÓDIGO V33 ---")
    print(f"   Trade no cerrado antes del fin del día.")


# =======================================================================
# --- EJECUCIÓN DEL ANÁLISIS SOLICITADO ---
# =======================================================================

# DÍA CLAVE A ANALIZAR: 2025-10-10
FECHA_A_DEPURAR = '2025-10-10'

# Asume que df_nq_1m, calculate_orb_1m_v7_clean, y check_sequential_ohlc_olhc_v33 
# están definidas en celdas anteriores.

simulate_orb_trade_detail(
    FECHA_A_DEPURAR, 
    df_nq_1m, 
    check_sequential_ohlc_olhc_v33, 
    calculate_orb_1m_v7_clean
)
```

```python
import pandas as pd

# =======================================================================
# --- FUNCIÓN DE SIMULACIÓN Y REPORTE DETALLADO (Muestra Datos ORB) ---
# =======================================================================

def simulate_orb_trade_detail(fecha_analisis, df_nq_1m, check_func, orb_calc_func):
    """
    Simula el trade mostrando los datos de las velas usadas para el cálculo del ORB.
    """
    
    print(f"==================================================")
    print(f"| INICIANDO ANÁLISIS DETALLADO: {fecha_analisis} |")
    print(f"==================================================")

    try:
        # 1. Filtrar datos del día
        df_day = df_nq_1m.loc[fecha_analisis].copy()
        
        # --- 1a. Mostrar datos de las velas 09:30:00 a 09:35:00 ---
        orb_data_check = df_day.between_time('09:30:00', '09:35:00')
        print("\n--- DATOS DE LAS VELAS DEL ORB (09:30:00 - 09:35:00) ---")
        if not orb_data_check.empty:
            print(orb_data_check[['Open', 'High', 'Low', 'Close']])
        else:
            print("❌ No se encontraron datos para 09:30:00 - 09:35:00.")

        # 2. Calcular Niveles ORB
        orb_levels_tuple = orb_calc_func(df_day)
        
    except KeyError:
        print(f"❌ Error: La fecha {fecha_analisis} no se encuentra o las funciones no están disponibles.")
        return

    # Desempaquetar tupla de 5 elementos (H, L, M, RT, Valid)
    try:
        orb_h, orb_l, orb_m, range_total, is_invalid = orb_levels_tuple
        if orb_h is None: raise ValueError("ORB inválido/vacío.")
        r_unit = range_total / 2 # Calcular Riesgo Unitario
    except ValueError as e:
        print(f"❌ Error de cálculo del ORB: {e}")
        return

    # 3. Detección de Breakout (09:36 a 10:00)
    df_breakout_check = df_day.between_time('09:36:00', '10:00:00')
    
    breakout_info = {'time': None, 'direction': None}
    
    # Encontrar el primer breakout (Long o Short)
    for index, row in df_breakout_check.iterrows():
        if row['High'] >= orb_h:
            if breakout_info['direction'] is None:
                breakout_info = {'time': index, 'direction': 'Long'}
            break
        if row['Low'] <= orb_l:
            if breakout_info['direction'] is None:
                breakout_info = {'time': index, 'direction': 'Short'}
            break
    
    # 4. Asignar Trade
    direction = breakout_info['direction']
    if direction == 'Short':
        entry_price = orb_m
        sl_price = orb_h
        tp_price = entry_price - (2 * r_unit)
        be_activation_level = entry_price - r_unit
        retest_level = entry_price
        df_retest_check = df_day.loc[breakout_info['time']:].between_time('09:36:00', '10:00:00')
        retest_touch = df_retest_check[df_retest_check['High'] >= retest_level].head(1)
        
    elif direction == 'Long':
        entry_price = orb_m
        sl_price = orb_l
        tp_price = entry_price + (2 * r_unit)
        be_activation_level = entry_price + r_unit
        retest_level = entry_price
        df_retest_check = df_day.loc[breakout_info['time']:].between_time('09:36:00', '10:00:00')
        retest_touch = df_retest_check[df_retest_check['Low'] <= retest_level].head(1)
    else:
        print("⛔ No se encontró un Breakout válido entre 09:36 y 10:00. Fin del análisis.")
        return

    print("\n--- 1. NIVELES Y DIRECCIÓN ---")
    print(f"   BREAKOUT DETECTADO: {breakout_info['direction']} en {breakout_info['time'].strftime('%H:%M:%S') if breakout_info['time'] else 'N/A'}")
    print(f"   Dirección del Trade: {direction}")
    print(f"   Entrada (ORB Mid):  {entry_price:.2f}")
    print(f"   SL Inicial (1R):    {sl_price:.2f}")
    print(f"   BE Activación (1R): {be_activation_level:.2f}")
    print(f"   TP (2R):            {tp_price:.2f}")
    
    if retest_touch.empty:
        print(f"⛔ No hubo retest (entrada) después del breakout {direction} antes de las 10:00. Fin del análisis.")
        return
    
    entry_time = retest_touch.index[0]
    print(f"\n   ✅ Entrada (Retest) detectada en: {entry_time.strftime('%H:%M:%S')}")


    # 5. Simular la Ejecución Vela a Vela (Omitido por el error de Breakout)
    print("\n--- 2. SEGUIMIENTO VELA A VELA ---")
    print("⚠️ La simulación de ejecución detallada se detiene aquí para enfocarnos en la corrección del Breakout FALSO POSITIVO.")

# =======================================================================
# --- EJECUCIÓN DEL ANÁLISIS SOLICITADO ---
# =======================================================================

FECHA_A_DEPURAR = '2025-10-10'

simulate_orb_trade_detail(
    FECHA_A_DEPURAR, 
    df_nq_1m, 
    check_sequential_ohlc_olhc_v33, 
    calculate_orb_1m_v7_clean
)
```

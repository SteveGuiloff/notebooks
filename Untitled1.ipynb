```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt


def run_montecarlo(csv_text, iterations=10000, initial_capital=50000):
    # Pre-procesamiento
    df = pd.read_csv(io.StringIO(csv_text))
    
    # Filtrar solo líneas de salida (Exit) para evitar duplicados y extraer P&L
    pnl_array = df[df['Type'].str.contains('Exit', na=False)]['Net P&L USD'].values
    
    if len(pnl_array) == 0:
        return "No se encontraron trades de salida."

    final_profits = []
    max_drawdowns = []
    ruin_count = 0
    limit_ruin = initial_capital * 0.5 # 50% de la cuenta
    
    # Simulación
    for i in range(iterations):
        # Reordenar con reemplazo (Bootstrap)
        sim_pnl = np.random.choice(pnl_array, size=len(pnl_array), replace=True)
        
        # Curva de equidad
        equity_curve = initial_capital + np.cumsum(sim_pnl)
        equity_curve = np.insert(equity_curve, 0, initial_capital)
        
        # Profit Final
        final_profits.append(equity_curve[-1] - initial_capital)
        
        # Drawdown
        peak = np.maximum.accumulate(equity_curve)
        dd = (peak - equity_curve)
        max_drawdowns.append(np.max(dd))
        
        # Riesgo de Ruina
        if np.min(equity_curve) <= (initial_capital - limit_ruin):
            ruin_count += 1
            
    # Resultados
    results = {
        "mean_profit": np.mean(final_profits),
        "median_profit": np.median(final_profits),
        "prob_positive": (np.array(final_profits) > 0).mean() * 100,
        "avg_max_dd": np.mean(max_drawdowns),
        "dd_95": np.percentile(max_drawdowns, 95),
        "dd_99": np.percentile(max_drawdowns, 99),
        "prob_ruin": (ruin_count / iterations) * 100,
        "total_trades": len(pnl_array)
    }
    
    return results

# Ejecutar
stats = run_montecarlo(trades_data)

# Imprimir Informe
print("Resultados Montecarlo (10,000 runs):")
print(f"------------------------------------")
print(f"Muestra: {stats['total_trades']} trades.")
print(f"Profit Promedio: ${stats['mean_profit']:.2f}")
print(f"Mediana Profit: ${stats['median_profit']:.2f}")
print(f"Probabilidad de terminar en positivo: {stats['prob_positive']:.2f}%")
print(f"------------------------------------")
print(f"Max DD (Media): ${stats['avg_max_dd']:.2f} ({ (stats['avg_max_dd']/50000)*100:.2f}%)")
print(f"Max DD (95% Conf.): ${stats['dd_95']:.2f}")
print(f"Max DD (99% Conf.): ${stats['dd_99']:.2f}")
print(f"------------------------------------")
print(f"Riesgo de Ruina (Pérdida 50% capital): {stats['prob_ruin']:.2f}%")
```

```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt

# --- CONFIGURACIÓN GLOBAL DEL BREAKER ---
WINDOW_SIZE = 10      # Ventana móvil de trades
THRESHOLD_OFF = 0.8   # PF por debajo del cual se apaga la operativa (Modo Simulado)
THRESHOLD_ON = 1.2    # PF por encima del cual se reactiva la operativa (Modo Real)



def process_breaker_system(csv_text, window, threshold_off, threshold_on):
    # --- PRE-PROCESAMIENTO ---
    df_raw = pd.read_csv(io.StringIO(csv_text))
    
    # Filtrar solo Exits y columnas necesarias
    df = df_raw[df_raw['Type'].str.contains('Exit', na=False)].copy()
    df['Date'] = pd.to_datetime(df['Date and time'])
    df['Net_PL'] = pd.to_numeric(df['Net P&L USD'])
    df['Is_Winner'] = df['Net_PL'] > 0
    df = df.sort_values('Date').reset_index(drop=True)
    
    # --- MÉTRICAS DE ROLLING PF ---
    def calculate_pf(series):
        wins = series[series > 0].sum()
        losses = abs(series[series < 0].sum())
        if losses == 0: return 10.0 # Valor techo para rachas ganadoras
        return wins / losses

    df['Rolling_PF'] = df['Net_PL'].rolling(window=window, min_periods=1).apply(calculate_pf)
    
    # --- LÓGICA DEL BREAKER (Recorrido secuencial) ---
    system_active = True
    breaker_pnl = []
    status_log = []
    
    for i in range(len(df)):
        # Decidimos basándonos en el PF del trade anterior para evitar look-ahead bias
        # Si es el primer trade, asumimos sistema activo (PF inicial arbitrario > threshold_on)
        current_pf = df['Rolling_PF'].iloc[i-1] if i > 0 else 1.5 
        
        # Lógica de cambio de estado
        if system_active and current_pf < threshold_off:
            system_active = False
        elif not system_active and current_pf > threshold_on:
            system_active = True
        
        status_log.append(system_active)
        breaker_pnl.append(df['Net_PL'].iloc[i] if system_active else 0.0)
    
    df['Breaker_PL'] = breaker_pnl
    df['System_Active'] = status_log
    
    # --- CURVAS Y DRAWDOWN ---
    df['Equity_Original'] = df['Net_PL'].cumsum()
    df['Equity_Breaker'] = df['Breaker_PL'].cumsum()
    
    def get_dd(equity_series):
        peak = equity_series.cummax()
        return (peak - equity_series)

    df['DD_Original'] = get_dd(df['Equity_Original'])
    df['DD_Breaker'] = get_dd(df['Equity_Breaker'])
    
    # --- PROFIT FACTOR TOTAL ---
    def get_total_pf(pnl_series):
        gross_profit = pnl_series[pnl_series > 0].sum()
        gross_loss = abs(pnl_series[pnl_series < 0].sum())
        return gross_profit / gross_loss if gross_loss > 0 else 10.0

    total_pf_original = get_total_pf(df['Net_PL'])
    total_pf_breaker = get_total_pf(df['Breaker_PL'])
    
    return df, total_pf_original, total_pf_breaker

# Ejecución usando las constantes globales
results_df, pf_orig, pf_break = process_breaker_system(trades_data, WINDOW_SIZE, THRESHOLD_OFF, THRESHOLD_ON)

# --- VISUALIZACIÓN ---
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 14), sharex=True)

# Gráfico 1: Equity
ax1.plot(results_df.index, results_df['Equity_Original'], label='Equity Original (Full)', color='gray', alpha=0.5, linestyle='--')
ax1.plot(results_df.index, results_df['Equity_Breaker'], label='Equity con Breaker', color='#2962FF', linewidth=2)
ax1.set_title("Curva de Capital: Estrategia vs Breaker")
ax1.legend()
ax1.grid(True, alpha=0.2)

# Gráfico 2: Rolling PF
ax2.plot(results_df.index, results_df['Rolling_PF'], color='#7b1fa2', label=f'Rolling PF ({WINDOW_SIZE})')
ax2.axhline(THRESHOLD_OFF, color='#d32f2f', linestyle=':', label=f'Threshold OFF ({THRESHOLD_OFF})')
ax2.axhline(THRESHOLD_ON, color='#388e3c', linestyle=':', label=f'Threshold ON ({THRESHOLD_ON})')
ax2.set_title("Indicador de Desempeño: Rolling Profit Factor")
ax2.legend()
ax2.grid(True, alpha=0.2)

# Gráfico 3: Drawdown
ax3.fill_between(results_df.index, 0, results_df['DD_Original'], color='#f44336', alpha=0.15, label='DD Original')
ax3.fill_between(results_df.index, 0, results_df['DD_Breaker'], color='#4caf50', alpha=0.3, label='DD Breaker')
ax3.set_title("Comparativa de Drawdown (Puntos)")
ax3.invert_yaxis()
ax3.legend()
ax3.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()

# --- INFORME DE SALIDA ---
last_trade = results_df.iloc[-1]
avoided = (results_df['Breaker_PL'] == 0).sum()

print("Informe de Performance del Breaker:")
print(f"------------------------------------")
print(f"Profit Final (Sin Breaker): ${results_df['Equity_Original'].iloc[-1]:.2f}")
print(f"Profit Final (Con Breaker): ${results_df['Equity_Breaker'].iloc[-1]:.2f}")
print(f"------------------------------------")
print(f"Profit Factor TOTAL (Sin Breaker): {pf_orig:.4f}")
print(f"Profit Factor TOTAL (Con Breaker): {pf_break:.4f}")
print(f"------------------------------------")
print(f"Max DD (Sin Breaker): ${results_df['DD_Original'].max():.2f}")
print(f"Max DD (Con Breaker): ${results_df['DD_Breaker'].max():.2f}")
print(f"------------------------------------")
print(f"Número de Trades Evitados: {avoided}")
print(f"Estado Actual del Sistema: {'ACTIVO' if last_trade['System_Active'] else 'SIMULADO (OFF)'}")
print(f"Valor Rolling PF ({WINDOW_SIZE}) al último trade: {last_trade['Rolling_PF']:.4f}")
```

```python
import pandas as pd
import numpy as np
from datetime import time

# --- CONFIGURACIÓN DE ESCENARIO ---
# 'tp_friendly': Prioridad absoluta al TP si ambos se tocan (Optimista)
# 'sl_friendly': Prioridad absoluta al SL si ambos se tocan (Conservador)
# 'ohlc_logic':  Usa la dirección de la vela para inferir qué se tocó primero (Equilibrado)
SIM_MODE = 'sl_friendly' 

# --- CONSTANTES Y PARÁMETROS ---
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'
START_MONTH = '2025-12-01'
END_MONTH   = '2025-12-31'

TRADE_START = time(9, 30)
TRADE_END   = time(10, 30) 

PARAMS = {
    'risk_usd': 2000,
    'pt_value': 100,
    'comm': 5.2,          
    'slippage_pts': 0.1,  # 1 tick por ejecución
    'reward_ratio': 1.0,
    'max_risk_pts': 150.0,
    'qc_threshold': 0.8,
    'adx_min': 23,
    'min_atr': 2.5,
}

def wilders_smoothing(series, period):
    return series.ewm(alpha=1/period, adjust=False).mean()

try:
    print(f"⏳ Backtest Dic 2025 | Modo: {SIM_MODE} | Slippage: 0.1 pts")
    
    df_raw = pd.read_parquet(FILE_PATH)
    df_raw.index = df_raw.index + pd.Timedelta(hours=5) 
    
    df = df_raw.loc['2025-11-20':END_MONTH].copy()
    for col in ['Open', 'High', 'Low', 'Close']:
        adj_col = f"{col}_adj"
        df[col.lower()] = df[adj_col] if adj_col in df.columns else df[col]

    # Indicadores
    df['ema_11'] = df['close'].ewm(span=11, adjust=False).mean()
    df['ema_75'] = df['close'].ewm(span=75, adjust=False).mean()
    tr = pd.concat([df['high'] - df['low'], abs(df['high'] - df['close'].shift()), abs(df['low'] - df['close'].shift())], axis=1).max(axis=1)
    df['atr'] = wilders_smoothing(tr, 14)
    up, down = df['high'].diff(), -df['low'].diff()
    p_dm = np.where((up > down) & (up > 0), up, 0)
    m_dm = np.where((down > up) & (down > 0), down, 0)
    tr_s = wilders_smoothing(tr, 14)
    p_di = 100 * wilders_smoothing(pd.Series(p_dm, index=df.index), 14) / tr_s
    m_di = 100 * wilders_smoothing(pd.Series(m_dm, index=df.index), 14) / tr_s
    dx = 100 * abs(p_di - m_di) / (p_di + m_di)
    df['adx'] = wilders_smoothing(dx, 14)

    df_sim = df.loc[START_MONTH:END_MONTH].copy()
    all_trades = []
    
    for dia, group in df_sim.groupby(df_sim.index.date):
        data = group.reset_index().to_dict('records')
        is_active, pos = False, {}
        
        for i in range(2, len(data) - 1): 
            row, v1, v2, next_row = data[i], data[i-1], data[i-2], data[i+1]
            c_time = row['timestamp'].time()
            
            if is_active:
                exit_trade = False
                hit_sl = row['low'] <= pos['sl'] if pos['side'] == 1 else row['high'] >= pos['sl']
                hit_tp = row['high'] >= pos['tp'] if pos['side'] == 1 else row['low'] <= pos['tp']
                
                # LÓGICA DE RESOLUCIÓN DE CONFLICTOS (Intra-bar)
                if hit_sl and hit_tp:
                    if SIM_MODE == 'tp_friendly':
                        exit_type = "TP"
                    elif SIM_MODE == 'sl_friendly':
                        exit_type = "SL"
                    else: # OHLC Logic: Si la vela es a favor de la posición, asumimos TP
                        is_bullish = row['close'] > row['open']
                        exit_type = "TP" if (pos['side'] == 1 and is_bullish) or (pos['side'] == -1 and not is_bullish) else "SL"
                elif hit_tp:
                    exit_type = "TP"
                elif hit_sl:
                    exit_type = "SL"
                else:
                    exit_type = None

                if exit_type:
                    pnl_pts = (pos['tp'] - pos['entry'] if exit_type == "TP" else pos['sl'] - pos['entry']) * pos['side']
                    pnl_pts -= (PARAMS['slippage_pts'] * 2) # Slippage entrada + salida
                    net_pnl = (pnl_pts * PARAMS['pt_value'] * pos['qty']) - (pos['qty'] * PARAMS['comm'] * 2)
                    all_trades.append({'Fecha': dia, 'Señal': pos['signal_time'], 'Tipo': exit_type, 'Qty': pos['qty'], 'Pts': round(pnl_pts, 2), 'PnL': round(net_pnl, 2)})
                    is_active = False
                elif c_time >= time(15, 56):
                    pnl_pts = ((row['close'] - pos['entry']) * pos['side']) - (PARAMS['slippage_pts'] * 2)
                    net_pnl = (pnl_pts * PARAMS['pt_value'] * pos['qty']) - (pos['qty'] * PARAMS['comm'] * 2)
                    all_trades.append({'Fecha': dia, 'Señal': pos['signal_time'], 'Tipo': "Close", 'Qty': pos['qty'], 'Pts': round(pnl_pts, 2), 'PnL': round(net_pnl, 2)})
                    is_active = False
                continue

            if TRADE_START <= c_time < TRADE_END:
                range_v = (row['high'] - row['low'])
                qc = (row['close'] - row['low']) / range_v if range_v > 0 else 0.5
                momentum_ok = row['adx'] >= PARAMS['adx_min'] and row['atr'] >= PARAMS['min_atr']
                
                # SHORT
                if not is_active and v1['low'] > v2['low'] and v1['high'] > v2['high'] and v1['close'] < v1['open'] and row['close'] < v1['low'] and qc <= (1 - PARAMS['qc_threshold']) and momentum_ok:
                    if row['close'] < row['ema_11'] or row['close'] < row['ema_75']:
                        risk_pts = v2['high'] - row['close']
                        if 0 < risk_pts <= PARAMS['max_risk_pts']:
                            qty = min(int(max(np.floor(PARAMS['risk_usd'] / (risk_pts * PARAMS['pt_value'])), 1)), 8)
                            is_active, pos = True, {'entry': next_row['open'], 'side': -1, 'qty': qty, 'sl': next_row['open'] + risk_pts, 'tp': next_row['open'] - (risk_pts * PARAMS['reward_ratio']), 'signal_time': row['timestamp'].strftime('%H:%M')}
                
                # LONG
                if not is_active and v1['high'] < v2['high'] and v1['low'] < v2['low'] and v1['close'] > v1['open'] and row['close'] > v1['high'] and qc >= PARAMS['qc_threshold'] and momentum_ok:
                    if row['close'] > row['ema_11'] or row['close'] > row['ema_75']:
                        risk_pts = row['close'] - v2['low']
                        if 0 < risk_pts <= PARAMS['max_risk_pts']:
                            qty = min(int(max(np.floor(PARAMS['risk_usd'] / (risk_pts * PARAMS['pt_value'])), 1)), 8)
                            is_active, pos = True, {'entry': next_row['open'], 'side': 1, 'qty': qty, 'sl': next_row['open'] - risk_pts, 'tp': next_row['open'] + (risk_pts * PARAMS['reward_ratio']), 'signal_time': row['timestamp'].strftime('%H:%M')}

    df_res = pd.DataFrame(all_trades)
    if not df_res.empty:
        net_pnl = df_res['PnL'].sum()
        gross_p, gross_l = df_res[df_res['PnL'] > 0]['PnL'].sum(), abs(df_res[df_res['PnL'] < 0]['PnL'].sum())
        print(df_res.to_string(index=False))
        print(f"\nNet PnL: ${net_pnl:,.2f} | PF: {gross_p/gross_l:.3f} | WR: {(len(df_res[df_res['PnL']>0])/len(df_res)*100):.2f}%")

except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import pandas as pd
import numpy as np
from datetime import time

# --- 1. CONFIGURACIÓN DEL MOTOR ---
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'
START_DATE = '2025-12-01'
END_DATE   = '2025-12-31'

class MicrostructureBacktester:
    def __init__(self, file_path, pt_value=100, comm=5.2):
        print("🚀 Cargando y sincronizando datos...")
        self.df = pd.read_parquet(file_path)
        self.df.columns = [c.strip() for c in self.df.columns]
        
        # Sincronización horaria confirmada (+5h para NY)
        self.df.index = self.df.index + pd.Timedelta(hours=5)
        
        # Ordenar índice para asegurar cálculos temporales correctos
        self.df = self.df.sort_index()
        
        self.pt_value = pt_value
        self.comm = comm
        self.results = []

    def calculate_indicators(self, df):
        """Cálculos técnicos espejados de Pine Script (Wilder's Smoothing)"""
        # EMAs estándar
        df['ema_75'] = df['Close'].ewm(span=75, adjust=False).mean()
        df['ema_11'] = df['Close'].ewm(span=11, adjust=False).mean()
        
        # ATR (Wilder's Smoothing - Esencial para coincidir con TV)
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        
        # En Pine, ta.atr(14) usa RMA (Moving Average de Wilder)
        # alpha = 1/periodo es la clave de Wilder
        df['atr'] = tr.ewm(alpha=1/14, min_periods=1, adjust=False).mean()
        
        # ADX (DMI de Wilder)
        up = df['High'].diff()
        down = -df['Low'].diff()
        
        plus_dm = np.where((up > down) & (up > 0), up, 0)
        minus_dm = np.where((down > up) & (down > 0), down, 0)
        
        # Suavizado RMA para DMI
        tr_rma = tr.ewm(alpha=1/14, min_periods=1, adjust=False).mean()
        p_dm_smooth = pd.Series(plus_dm, index=df.index).ewm(alpha=1/14, min_periods=1, adjust=False).mean()
        m_dm_smooth = pd.Series(minus_dm, index=df.index).ewm(alpha=1/14, min_periods=1, adjust=False).mean()
        
        plus_di = 100 * p_dm_smooth / tr_rma
        minus_di = 100 * m_dm_smooth / tr_rma
        
        dx = 100 * np.abs(plus_di - minus_di) / (plus_di + minus_di).replace(0, np.nan)
        df['adx'] = dx.ewm(alpha=1/14, min_periods=1, adjust=False).mean()
        
        return df

    def run(self, start_date, end_date, params):
        # IMPORTANTE: Calculamos indicadores sobre TODO el dataset primero
        # para tener datos de "warm-up" antes de Diciembre.
        df_full = self.calculate_indicators(self.df.copy())
        
        # Ahora filtramos por el periodo de interés
        df = df_full.loc[start_date:end_date].copy()
        
        if df.empty:
            print(f"⚠️ El DataFrame filtrado está vacío para el rango {start_date} - {end_date}")
            return pd.DataFrame()

        # Preparación para iteración rápida
        data = df.reset_index().to_dict('records')
        self.results = []
        
        is_active = False
        side = 0 # 1 Long, -1 Short
        entry_p, sl_p, tp_p, current_risk = 0, 0, 0, 0
        qty = 0
        be_active = False

        print(f"📈 Simulando periodo: {start_date} a {end_date}")

        for i in range(2, len(data)):
            row = data[i]
            v1 = data[i-1]
            v2 = data[i-2]
            
            # --- GESTIÓN DE SALIDAS ---
            if is_active:
                if side == 1:
                    if row['Low'] <= sl_p:
                        self._add_trade(row['timestamp'], sl_p - entry_p, qty, "SL/BE")
                        is_active = False
                    elif row['High'] >= tp_p:
                        self._add_trade(row['timestamp'], tp_p - entry_p, qty, "TP")
                        is_active = False
                else:
                    if row['High'] >= sl_p:
                        self._add_trade(row['timestamp'], entry_p - sl_p, qty, "SL/BE")
                        is_active = False
                    elif row['Low'] <= tp_p:
                        self._add_trade(row['timestamp'], entry_p - tp_p, qty, "TP")
                        is_active = False

                if is_active and not be_active:
                    if (side == 1 and row['High'] >= entry_p + (current_risk * params['be_trigger'])) or \
                       (side == -1 and row['Low'] <= entry_p - (current_risk * params['be_trigger'])):
                        sl_p = entry_p + (params['be_offset'] * side)
                        be_active = True
                
                if is_active and row['timestamp'].time() >= time(15, 56):
                    pnl_pts = (row['Close'] - entry_p) * side
                    self._add_trade(row['timestamp'], pnl_pts, qty, "Forced")
                    is_active = False
                
                if not is_active: continue

            # --- LÓGICA DE ENTRADA ---
            curr_t = row['timestamp'].time()
            if not is_active and time(9, 30) <= curr_t <= time(11, 0):
                
                # Filtros de Volatilidad y Momentum
                vol_ok = not np.isnan(row['adx']) and row['adx'] >= params['adx_min'] and \
                         params['min_atr'] <= row['atr'] <= params['max_atr']
                
                # Calidad de Cierre
                rng = row['High'] - row['Low']
                qc = (row['Close'] - row['Low']) / rng if rng != 0 else 0.5
                
                # SHORT SIGNAL
                short_pause = v1['Low'] > v2['Low'] and v1['High'] > v2['High']
                short_v2_ok = v1['Close'] < v1['Open']
                short_trigger = row['Close'] < v1['Low']
                
                if short_v2_ok and short_pause and short_trigger and vol_ok:
                    if qc <= (1 - params['qc_threshold']):
                        if row['Close'] < row['ema_75'] or row['Close'] < row['ema_11']:
                            side = -1
                            current_risk = v2['High'] - row['Close']
                            if 0 < current_risk <= params['max_risk_pts']:
                                qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                                if qty >= 1:
                                    is_active, entry_p = True, row['Close']
                                    sl_p, tp_p = entry_p + current_risk, entry_p - (current_risk * params['reward_ratio'])
                                    be_active = False

                # LONG SIGNAL
                long_pause = v1['High'] < v2['High'] and v1['Low'] < v2['Low']
                long_v2_ok = v1['Close'] > v1['Open']
                long_trigger = row['Close'] > v1['High']
                
                if long_v2_ok and long_pause and long_trigger and vol_ok:
                    if qc >= params['qc_threshold']:
                        if row['Close'] > row['ema_75'] or row['Close'] > row['ema_11']:
                            side = 1
                            current_risk = row['Close'] - v2['Low']
                            if 0 < current_risk <= params['max_risk_pts']:
                                qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                                if qty >= 1:
                                    is_active, entry_p = True, row['Close']
                                    sl_p, tp_p = entry_p - current_risk, entry_p + (current_risk * params['reward_ratio'])
                                    be_active = False

        return pd.DataFrame(self.results)

    def _add_trade(self, time, points, qty, exit_type):
        pnl = (points * qty * self.pt_value) - (self.comm * 2 * qty)
        self.results.append({'time': time, 'points': points, 'pnl': pnl, 'qty': qty, 'type': exit_type})

# --- PARÁMETROS DE PRODUCCIÓN ---
params = {
    'adx_min': 23, 
    'min_atr': 2.5, 
    'max_atr': 15.0, 
    'qc_threshold': 0.8,
    'risk_usd': 2000, 
    'max_contracts': 8, 
    'reward_ratio': 1.0, 
    'be_trigger': 1.5, 
    'be_offset': 1.0, 
    'max_risk_pts': 150.0
}

tester = MicrostructureBacktester(FILE_PATH)
trades = tester.run(START_DATE, END_DATE, params)

if not trades.empty:
    print(f"\n✅ RESULTADOS DICIEMBRE 2025:")
    print(f"Total Trades: {len(trades)}")
    print(f"PnL Total: ${trades['pnl'].sum():,.2f}")
    print(f"Win Rate: {(trades['pnl'] > 0).mean():.2%}")
    # Manejo de división por cero en Profit Factor si no hay pérdidas
    losses = abs(trades[trades['pnl'] < 0]['pnl'].sum())
    pf = trades[trades['pnl'] > 0]['pnl'].sum() / losses if losses != 0 else float('inf')
    print(f"Profit Factor: {pf:.3f}")
else:
    print("\n❌ No se detectaron trades.")
    # Debug más profundo
    df_debug = tester.calculate_indicators(tester.df.copy()).loc[START_DATE:END_DATE]
    print(f"Registros en Diciembre: {len(df_debug)}")
    print(f"ADX Promedio: {df_debug['adx'].mean():.2f}")
    print(f"ATR Promedio: {df_debug['atr'].mean():.2f}")
    print(f"Cierres > EMA75: {(df_debug['Close'] > df_debug['ema_75']).sum()}")
```

```python
import pandas as pd
import numpy as np
from datetime import time
import matplotlib.pyplot as plt

# --- CONFIGURACIÓN ---
SIM_MODE = 'tp_friendly' 

# --- CONSTANTES Y PARÁMETROS ---
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'

# Ventana de trading (NY Time)
TRADE_START = time(9, 30)
TRADE_END   = time(10, 30) 

PARAMS = {
    'risk_usd': 2000,
    'pt_value': 100,
    'comm': 5.2,          
    'slippage_pts': 0.1,  
    'reward_ratio': 1.0,
    'max_risk_pts': 150.0,
    'qc_threshold': 0.8,
    'adx_min': 23,
    'min_atr': 2.5,
    'max_contracts': 8
}

def wilders_smoothing(series, period):
    """Replicación exacta de ta.rma (Wilder's Moving Average)"""
    return series.ewm(alpha=1/period, adjust=False).mean()

try:
    print(f"⏳ Iniciando Backtest Histórico Completo | Modo: {SIM_MODE}")
    
    # Carga de datos
    df_raw = pd.read_parquet(FILE_PATH)
    
    # Sincronización horaria (+5h para NY)
    df_raw.index = df_raw.index + pd.Timedelta(hours=5) 
    
    df = df_raw.copy()
    for col in ['Open', 'High', 'Low', 'Close']:
        adj_col = f"{col}_adj"
        df[col.lower()] = df[adj_col] if adj_col in df.columns else df[col]

    # --- CÁLCULO DE INDICADORES ---
    df['ema_11'] = df['close'].ewm(span=11, adjust=False).mean()
    df['ema_75'] = df['close'].ewm(span=75, adjust=False).mean()
    
    tr = pd.concat([df['high'] - df['low'], 
                    abs(df['high'] - df['close'].shift()), 
                    abs(df['low'] - df['close'].shift())], axis=1).max(axis=1)
    df['atr'] = wilders_smoothing(tr, 14)
    
    up, down = df['high'].diff(), -df['low'].diff()
    p_dm = np.where((up > down) & (up > 0), up, 0)
    m_dm = np.where((down > up) & (down > 0), down, 0)
    tr_s = wilders_smoothing(tr, 14)
    p_di = 100 * wilders_smoothing(pd.Series(p_dm, index=df.index), 14) / tr_s
    m_di = 100 * wilders_smoothing(pd.Series(m_dm, index=df.index), 14) / tr_s
    dx = 100 * abs(p_di - m_di) / (p_di + m_di)
    df['adx'] = wilders_smoothing(dx, 14)

    df_sim = df.dropna(subset=['adx', 'atr']).copy()
    
    all_trades = []
    
    # --- PROCESO DE SIMULACIÓN (Ejecución única) ---
    for dia, group in df_sim.groupby(df_sim.index.date):
        data = group.reset_index().to_dict('records')
        is_active, pos = False, {}
        
        for i in range(2, len(data) - 1): 
            row, v1, v2, next_row = data[i], data[i-1], data[i-2], data[i+1]
            c_time = row['timestamp'].time()
            
            if is_active:
                hit_sl = row['low'] <= pos['sl'] if pos['side'] == 1 else row['high'] >= pos['sl']
                hit_tp = row['high'] >= pos['tp'] if pos['side'] == 1 else row['low'] <= pos['tp']
                
                exit_type = None
                if hit_sl and hit_tp:
                    exit_type = "TP" if SIM_MODE == 'tp_friendly' else "SL"
                elif hit_tp: exit_type = "TP"
                elif hit_sl: exit_type = "SL"
                
                if exit_type:
                    pnl_pts = (pos['tp'] - pos['entry'] if exit_type == "TP" else pos['sl'] - pos['entry']) * pos['side']
                    pnl_pts -= (PARAMS['slippage_pts'] * 2) 
                    net_pnl = (pnl_pts * PARAMS['pt_value'] * pos['qty']) - (pos['qty'] * PARAMS['comm'] * 2)
                    
                    all_trades.append({
                        'Fecha': pd.to_datetime(dia), 
                        'Señal': pos['signal_time'], 
                        'Tipo': exit_type, 
                        'Side': 'Long' if pos['side'] == 1 else 'Short',
                        'Qty': pos['qty'], 
                        'Pts': round(pnl_pts, 2), 
                        'PnL': round(net_pnl, 2)
                    })
                    is_active = False
                elif c_time >= time(15, 56):
                    pnl_pts = ((row['close'] - pos['entry']) * pos['side']) - (PARAMS['slippage_pts'] * 2)
                    net_pnl = (pnl_pts * PARAMS['pt_value'] * pos['qty']) - (pos['qty'] * PARAMS['comm'] * 2)
                    all_trades.append({
                        'Fecha': pd.to_datetime(dia), 'Señal': pos['signal_time'], 'Tipo': "Forced", 
                        'Side': 'Long' if pos['side'] == 1 else 'Short',
                        'Qty': pos['qty'], 'Pts': round(pnl_pts, 2), 'PnL': round(net_pnl, 2)
                    })
                    is_active = False
                continue

            if TRADE_START <= c_time < TRADE_END:
                range_v = (row['high'] - row['low'])
                qc = (row['close'] - row['low']) / range_v if range_v > 0 else 0.5
                momentum_ok = row['adx'] >= PARAMS['adx_min'] and row['atr'] >= PARAMS['min_atr']
                
                if not is_active and v1['low'] > v2['low'] and v1['high'] > v2['high'] and v1['close'] < v1['open'] and row['close'] < v1['low'] and qc <= (1 - PARAMS['qc_threshold']) and momentum_ok:
                    if row['close'] < row['ema_11'] or row['close'] < row['ema_75']:
                        risk_pts = v2['high'] - row['close']
                        if 0 < risk_pts <= PARAMS['max_risk_pts']:
                            qty = min(int(max(np.floor(PARAMS['risk_usd'] / (risk_pts * PARAMS['pt_value'])), 1)), PARAMS['max_contracts'])
                            is_active, pos = True, {
                                'entry': next_row['open'], 'side': -1, 'qty': qty,
                                'sl': next_row['open'] + risk_pts, 'tp': next_row['open'] - (risk_pts * PARAMS['reward_ratio']),
                                'signal_time': row['timestamp'].strftime('%H:%M')
                            }
                
                if not is_active and v1['high'] < v2['high'] and v1['low'] < v2['low'] and v1['close'] > v1['open'] and row['close'] > v1['high'] and qc >= PARAMS['qc_threshold'] and momentum_ok:
                    if row['close'] > row['ema_11'] or row['close'] > row['ema_75']:
                        risk_pts = row['close'] - v2['low']
                        if 0 < risk_pts <= PARAMS['max_risk_pts']:
                            qty = min(int(max(np.floor(PARAMS['risk_usd'] / (risk_pts * PARAMS['pt_value'])), 1)), PARAMS['max_contracts'])
                            is_active, pos = True, {
                                'entry': next_row['open'], 'side': 1, 'qty': qty,
                                'sl': next_row['open'] - risk_pts, 'tp': next_row['open'] + (risk_pts * PARAMS['reward_ratio']),
                                'signal_time': row['timestamp'].strftime('%H:%M')
                            }

    # --- REPORTE Y GRÁFICOS USANDO df_res ---
    df_res = pd.DataFrame(all_trades)
    
    if not df_res.empty:
        # 1. Métricas de Consola
        total_pnl = df_res['PnL'].sum()
        print(f"\n✅ Análisis Completado: {len(df_res)} trades encontrados.")
        print(f"PnL Total: ${total_pnl:,.2f}")
        
        # 2. Generar Curva de Equidad
        df_res['Equity'] = df_res['PnL'].cumsum()
        
        plt.figure(figsize=(12, 6))
        plt.plot(df_res['Equity'].values, color='#089981', linewidth=2, label='Equity Line')
        plt.fill_between(range(len(df_res)), df_res['Equity'].values, alpha=0.1, color='#089981')
        
        plt.title('Curva de Equidad Histórica - GC Microstructure (Basado en df_res)')
        plt.xlabel('Número de Operación')
        plt.ylabel('Balance Acumulado (USD)')
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print("❌ El DataFrame de resultados está vacío. No hay datos para graficar.")

except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import pandas as pd
import numpy as np
from datetime import time, timedelta

# --- CONSTANTES DE ESTRATEGIA ---
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'
TARGET_DATE = '2025-03-21'
MIN_ADX = 23.0
MIN_ATR = 2.5
PERIOD = 14  # Ambos indicadores usan 14 periodos

def wilders_smoothing(series, period):
    """
    Replicación exacta de ta.rma (Wilder's Smoothing).
    TradingView usa esta fórmula: RMA = (1/n) * Source + (1 - 1/n) * RMA[1]
    """
    return series.ewm(alpha=1/period, adjust=False).mean()

try:
    print(f"🕵️ Investigando discrepancia de indicadores (Periodo: {PERIOD})")
    
    # 1. CARGA DE DATOS CONTINUOS
    # Para que el ADX de 14 coincida con TV, necesitamos al menos 250 velas de pre-carga
    df_raw = pd.read_parquet(FILE_PATH)
    df_raw.columns = [c.capitalize() for c in df_raw.columns]
    df_raw.index = df_raw.index + pd.Timedelta(hours=5) # Sincronización NY
    
    target_dt = pd.to_datetime(TARGET_DATE)
    # Cargamos 15 días atrás para que la "memoria" de Wilder sea idéntica a la de TV
    df = df_raw.loc[target_dt - timedelta(days=15) : target_dt + timedelta(days=1)].copy()
    
    # 2. CÁLCULO DE INDICADORES (Lógica ta.atr y ta.adx de Pine Script)
    
    # --- ATR ---
    high_low = df['High'] - df['Low']
    high_close = abs(df['High'] - df['Close'].shift())
    low_close = abs(df['Low'] - df['Close'].shift())
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    df['atr_calc'] = wilders_smoothing(tr, PERIOD)
    
    # --- ADX ---
    up = df['High'].diff()
    down = -df['Low'].diff()
    
    plus_dm = np.where((up > down) & (up > 0), up, 0)
    minus_dm = np.where((down > up) & (down > 0), down, 0)
    
    tr_s = wilders_smoothing(tr, PERIOD)
    p_di = 100 * wilders_smoothing(pd.Series(plus_dm, index=df.index), PERIOD) / tr_s
    m_di = 100 * wilders_smoothing(pd.Series(minus_dm, index=df.index), PERIOD) / tr_s
    
    dx = 100 * abs(p_di - m_di) / (p_di + m_di)
    df['adx_calc'] = wilders_smoothing(dx, PERIOD)

    # 3. VERIFICACIÓN DE LOS TRADES FALTANTES
    df_day = df.loc[TARGET_DATE].copy()
    tiempos = ["09:48", "10:18"]
    
    print(f"\nResultados con Suavizado Wilder (n={PERIOD}):")
    print("-" * 90)
    print(f"{'Hora':<8} | {'Close':<8} | {'ADX (Py)':<10} | {'ATR (Py)':<10} | {'¿Cumple Filtros?'}")
    print("-" * 90)
    
    for t in tiempos:
        try:
            row = df_day.at_time(t).iloc[0]
            adx_val = row['adx_calc']
            atr_val = row['atr_calc']
            
            cumple = adx_val >= MIN_ADX and atr_val >= MIN_ATR
            status = "✅ SI" if cumple else "❌ NO"
            
            print(f"{t:<8} | {row['Close']:<8.1f} | {adx_val:<10.2f} | {atr_val:<10.2f} | {status}")
            
            if not cumple:
                if adx_val < MIN_ADX:
                    print(f"   ⚠️ FALLO ADX: {adx_val:.2f} < {MIN_ADX}")
                if atr_val < MIN_ATR:
                    print(f"   ⚠️ FALLO ATR: {atr_val:.2f} < {MIN_ATR}")
        except:
            print(f"{t:<8} | Sin datos.")

except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import pandas as pd
import numpy as np
from datetime import time

# --- 1. CONFIGURACIÓN DEL MOTOR ---
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'
START_DATE = '2025-03-01' 
END_DATE   = '2025-03-31'

class MicrostructureBacktester:
    def __init__(self, file_path, pt_value=100, comm=5.2):
        print("🚀 Cargando datos procesados y corregidos...")
        self.df = pd.read_parquet(file_path)
        
        if 'Timestamp_NY' in self.df.columns:
            self.df['Timestamp_NY'] = pd.to_datetime(self.df['Timestamp_NY'])
            self.df.set_index('Timestamp_NY', inplace=True)
        
        self.df = self.df.sort_index()
        
        # Usamos los precios AJUSTADOS para indicadores y ejecución (estándar de backtesting continuo)
        self.ohlc_cols = {
            'open': 'Open_adj',
            'high': 'High_adj',
            'low': 'Low_adj',
            'close': 'Close_adj'
        }
        
        self.pt_value = pt_value
        self.comm = comm
        self.results = []

    def calculate_indicators(self, df):
        """Cálculos técnicos con Wilder's Smoothing (RMA) sobre precios ajustados"""
        c = self.ohlc_cols['close']
        h = self.ohlc_cols['high']
        l = self.ohlc_cols['low']

        # EMAs
        df['ema_75'] = df[c].ewm(span=75, adjust=False).mean()
        df['ema_11'] = df[c].ewm(span=11, adjust=False).mean()
        
        # ATR (Wilder)
        hl = df[h] - df[l]
        hc = np.abs(df[h] - df[c].shift())
        lc = np.abs(df[l] - df[c].shift())
        tr = pd.concat([hl, hc, lc], axis=1).max(axis=1)
        df['atr'] = tr.ewm(alpha=1/14, adjust=False).mean()
        
        # ADX (Wilder)
        up = df[h].diff()
        down = -df[l].diff()
        p_dm = np.where((up > down) & (up > 0), up, 0)
        m_dm = np.where((down > up) & (down > 0), down, 0)
        
        tr_s = tr.ewm(alpha=1/14, adjust=False).mean()
        p_di = 100 * pd.Series(p_dm, index=df.index).ewm(alpha=1/14, adjust=False).mean() / tr_s
        m_di = 100 * pd.Series(m_dm, index=df.index).ewm(alpha=1/14, adjust=False).mean() / tr_s
        
        dx = 100 * np.abs(p_di - m_di) / (p_di + m_di).replace(0, np.nan)
        df['adx'] = dx.ewm(alpha=1/14, adjust=False).mean()
        
        return df

    def run(self, start_date, end_date, params):
        df_full = self.calculate_indicators(self.df.copy())
        df = df_full.loc[start_date:end_date].copy()
        
        if df.empty: 
            print(f"⚠️ No hay datos en el rango {start_date} a {end_date}")
            return pd.DataFrame()

        data = df.reset_index().to_dict('records')
        self.results = []
        
        is_active, side, entry_p, sl_p, tp_p, qty = False, 0, 0, 0, 0, 0
        current_risk, be_active = 0, False

        C, H, L, O = self.ohlc_cols['close'], self.ohlc_cols['high'], self.ohlc_cols['low'], self.ohlc_cols['open']

        for i in range(2, len(data)):
            row = data[i]
            v1, v2 = data[i-1], data[i-2]
            
            # --- Gestión de Salidas ---
            if is_active:
                if side == 1: # Largo
                    if row[L] <= sl_p:
                        self._add_trade(row['Timestamp_NY'], sl_p - entry_p, qty, "SL/BE")
                        is_active = False
                    elif row[H] >= tp_p:
                        self._add_trade(row['Timestamp_NY'], tp_p - entry_p, qty, "TP")
                        is_active = False
                else: # Corto
                    if row[H] >= sl_p:
                        self._add_trade(row['Timestamp_NY'], entry_p - sl_p, qty, "SL/BE")
                        is_active = False
                    elif row[L] <= tp_p:
                        self._add_trade(row['Timestamp_NY'], entry_p - tp_p, qty, "TP")
                        is_active = False

                if is_active and not be_active:
                    if (side == 1 and row[H] >= entry_p + (current_risk * params['be_trigger'])) or \
                       (side == -1 and row[L] <= entry_p - (current_risk * params['be_trigger'])):
                        sl_p = entry_p + (params['be_offset'] * side)
                        be_active = True
                
                if not is_active: continue

            # --- Lógica de Entrada (09:30 - 11:00 NY) ---
            curr_t = row['Timestamp_NY'].time()
            if not is_active and time(9, 30) <= curr_t <= time(11, 0):
                # Filtros de Volatilidad
                vol_ok = row['adx'] >= params['adx_min'] and \
                         params['min_atr'] <= row['atr'] <= params['max_atr']
                
                # Calidad de Cierre (QC)
                range_v = row[H] - row[L]
                qc_score = (row[C] - row[L]) / range_v if range_v > 0 else 0.5
                
                # SHORT
                if vol_ok and v1[L] > v2[L] and v1[H] > v2[H] and row[C] < v1[L]:
                    # Filtro de Cierre + Filtro de Tendencia (EMA 11 o 75)
                    if v1[C] < v1[O] and qc_score <= (1 - params['qc_threshold']):
                        if row[C] < row['ema_11'] or row[C] < row['ema_75']:
                            side, is_active = -1, True
                            entry_p = row[C]
                            current_risk = v2[H] - entry_p
                            qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                            sl_p, tp_p = entry_p + current_risk, entry_p - (current_risk * params['reward_ratio'])
                            be_active = False

                # LONG
                if vol_ok and v1[H] < v2[H] and v1[L] < v2[L] and row[C] > v1[H]:
                    if v1[C] > v1[O] and qc_score >= params['qc_threshold']:
                        if row[C] > row['ema_11'] or row[C] > row['ema_75']:
                            side, is_active = 1, True
                            entry_p = row[C]
                            current_risk = entry_p - v2[L]
                            qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                            sl_p, tp_p = entry_p - current_risk, entry_p + (current_risk * params['reward_ratio'])
                            be_active = False

        return pd.DataFrame(self.results)

    def _add_trade(self, time, points, qty, exit_type):
        pnl = (points * qty * self.pt_value) - (self.comm * 2 * qty)
        self.results.append({
            'time': time, 
            'points': round(points, 2), 
            'pnl': round(pnl, 2), 
            'qty': qty, 
            'type': exit_type
        })

# --- PARÁMETROS OPTIMIZADOS ---
params = {
    'adx_min': 20,           # Bajamos de 23 a 20 para capturar más momentum
    'min_atr': 2.5, 
    'max_atr': 15.0, 
    'qc_threshold': 0.7,     # Bajamos de 0.8 a 0.7 para ser menos restrictivos con la mecha
    'risk_usd': 2000, 
    'max_contracts': 8, 
    'reward_ratio': 1.0, 
    'be_trigger': 1.5, 
    'be_offset': 1.0
}

tester = MicrostructureBacktester(FILE_PATH)
trades = tester.run(START_DATE, END_DATE, params)

if not trades.empty:
    print(f"\n✅ BACKTEST FINALIZADO ({START_DATE} - {END_DATE}):")
    print(f"{'='*50}")
    print(f"Total Trades: {len(trades)}")
    print(f"PnL Neto:     ${trades['pnl'].sum():,.2f}")
    print(f"Win Rate:      {(trades['pnl'] > 0).mean():.2%}")
    neg_pnl = trades[trades['pnl'] < 0]['pnl'].sum()
    pf = abs(trades[trades['pnl'] > 0]['pnl'].sum() / neg_pnl) if neg_pnl != 0 else float('inf')
    print(f"Profit Factor: {pf:.2f}")
    print(f"{'='*50}\n")
    print(trades)
else:
    print("\n❌ Sin trades detectados con los nuevos parámetros.")
```

```python
import pandas as pd
import numpy as np
from datetime import time

# --- 1. CONFIGURACIÓN DEL MOTOR ---
# Usamos el dataset procesado por la pipeline final que ya incluye ajustes y hora NY
FILE_PATH = '/home/quant/data/processed/gc_1m_raw_continuous.parquet'
# Rango de validación: Diciembre 2025
START_DATE = '2025-12-01' 
END_DATE   = '2025-12-31'

class MicrostructureBacktester:
    def __init__(self, file_path, pt_value=100, comm=5.2):
        print("🚀 Cargando motor compatible con Dataset V2 (Ajustado y NY-Naive)...")
        self.df = pd.read_parquet(file_path)
        
        # El nuevo dataset ya tiene Timestamp_NY como índice o columna.
        # Si es columna, lo movemos al índice.
        if 'Timestamp_NY' in self.df.columns:
            self.df.set_index('Timestamp_NY', inplace=True)
        
        self.df.index = pd.to_datetime(self.df.index)
        self.df = self.df.sort_index()
        
        # Mapeo a columnas AJUSTADAS para evitar distorsión en indicadores
        self.ohlc_cols = {
            'open': 'Open_adj',
            'high': 'High_adj',
            'low': 'Low_adj',
            'close': 'Close_adj'
        }
        
        self.pt_value = pt_value
        self.comm = comm
        self.results = []

    def calculate_indicators(self, df):
        """Cálculos técnicos con Wilder's Smoothing (RMA) sobre precios ajustados"""
        c = self.ohlc_cols['close']
        h = self.ohlc_cols['high']
        l = self.ohlc_cols['low']

        # EMA rápida y lenta
        df['ema_75'] = df[c].ewm(span=75, adjust=False).mean()
        df['ema_11'] = df[c].ewm(span=11, adjust=False).mean()
        
        # ATR con RMA (Igual a TradingView)
        hl = df[h] - df[l]
        hc = np.abs(df[h] - df[c].shift())
        lc = np.abs(df[l] - df[c].shift())
        tr = pd.concat([hl, hc, lc], axis=1).max(axis=1)
        df['atr'] = tr.ewm(alpha=1/14, adjust=False).mean()
        
        # ADX con RMA
        up = df[h].diff()
        down = -df[l].diff()
        p_dm = np.where((up > down) & (up > 0), up, 0)
        m_dm = np.where((down > up) & (down > 0), down, 0)
        
        tr_s = tr.ewm(alpha=1/14, adjust=False).mean()
        p_di = 100 * pd.Series(p_dm, index=df.index).ewm(alpha=1/14, adjust=False).mean() / tr_s
        m_di = 100 * pd.Series(m_dm, index=df.index).ewm(alpha=1/14, adjust=False).mean() / tr_s
        
        dx = 100 * np.abs(p_di - m_di) / (p_di + m_di).replace(0, np.nan)
        df['adx'] = dx.ewm(alpha=1/14, adjust=False).mean()
        
        return df

    def run(self, start_date, end_date, params):
        # 1. Calcular indicadores sobre todo el dataset para evitar sesgo de inicio
        df_full = self.calculate_indicators(self.df.copy())
        
        # 2. Filtrar el rango de fechas para el backtest
        df = df_full.loc[start_date:end_date].copy()
        
        if df.empty: 
            print(f"⚠️ No hay datos en el rango {start_date} a {end_date}")
            return pd.DataFrame()

        data = df.reset_index().to_dict('records')
        # Buscamos el nombre de la columna del índice que reset_index() suele llamar 'index' o 'Timestamp_NY'
        time_col = 'Timestamp_NY' if 'Timestamp_NY' in data[0] else 'index'
        
        self.results = []
        is_active, side, entry_p, sl_p, tp_p, qty = False, 0, 0, 0, 0, 0
        current_risk, be_active = 0, False
        bars_held = 0

        C, H, L, O = self.ohlc_cols['close'], self.ohlc_cols['high'], self.ohlc_cols['low'], self.ohlc_cols['open']

        for i in range(2, len(data)):
            row = data[i]
            v1, v2 = data[i-1], data[i-2]
            ts = row[time_col]
            
            # Filtro de Feriados (Configurable)
            if ts.month == 12 and ts.day in [24, 25, 31]:
                continue

            # --- GESTIÓN DE SALIDAS ---
            if is_active:
                bars_held += 1
                if side == 1: # LARGO
                    if row[L] <= sl_p:
                        self._add_trade(ts, sl_p - entry_p, qty, "SL/BE")
                        is_active = False
                    elif row[H] >= tp_p:
                        self._add_trade(ts, tp_p - entry_p, qty, "TP")
                        is_active = False
                else: # CORTO
                    if row[H] >= sl_p:
                        self._add_trade(ts, entry_p - sl_p, qty, "SL/BE")
                        is_active = False
                    elif row[L] <= tp_p:
                        self._add_trade(ts, entry_p - tp_p, qty, "TP")
                        is_active = False

                # Lógica de Break Even
                if is_active and not be_active:
                    if (side == 1 and row[H] >= entry_p + (current_risk * params['be_trigger'])) or \
                       (side == -1 and row[L] <= entry_p - (current_risk * params['be_trigger'])):
                        sl_p = entry_p + (params['be_offset'] * side)
                        be_active = True

                # Expiración por tiempo
                if is_active and bars_held >= params['max_bars_open']:
                    pnl_pts = (row[C] - entry_p) if side == 1 else (entry_p - row[C])
                    self._add_trade(ts, pnl_pts, qty, "Time Exit")
                    is_active = False
                
                if not is_active: continue

            # --- LÓGICA DE ENTRADA (Horario NY directo del index) ---
            curr_t = ts.time()
            if not is_active and time(9, 30) <= curr_t <= time(11, 0):
                vol_ok = row['adx'] >= params['adx_min'] and \
                         params['min_atr'] <= row['atr'] <= params['max_atr']
                
                range_v = row[H] - row[L]
                qc_score = (row[C] - row[L]) / range_v if range_v > 0 else 0.5
                
                # SEÑAL SHORT (Basada en estructura de micro-pausa)
                if vol_ok and v1[L] > v2[L] and v1[H] > v2[H] and row[C] < v1[L]:
                    if v1[C] < v1[O] and qc_score <= (1 - params['qc_threshold']):
                        if row[C] < row['ema_11'] or row[C] < row['ema_75']:
                            side, is_active = -1, True
                            entry_p = row[C]
                            current_risk = v2[H] - entry_p
                            qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                            sl_p, tp_p = entry_p + current_risk, entry_p - (current_risk * params['reward_ratio'])
                            be_active, bars_held = False, 0

                # SEÑAL LONG (Basada en estructura de micro-pausa)
                if not is_active and vol_ok and v1[H] < v2[H] and v1[L] < v2[L] and row[C] > v1[H]:
                    if v1[C] > v1[O] and qc_score >= params['qc_threshold']:
                        if row[C] > row['ema_11'] or row[C] > row['ema_75']:
                            side, is_active = 1, True
                            entry_p = row[C]
                            current_risk = entry_p - v2[L]
                            qty = min(int(params['risk_usd'] / (current_risk * self.pt_value)), params['max_contracts'])
                            sl_p, tp_p = entry_p - current_risk, entry_p + (current_risk * params['reward_ratio'])
                            be_active, bars_held = False, 0

        return pd.DataFrame(self.results)

    def _add_trade(self, time, points, qty, exit_type):
        if qty <= 0: return
        pnl = (points * qty * self.pt_value) - (self.comm * 2 * qty)
        self.results.append({
            'time': time, 
            'points': round(points, 2), 
            'pnl': round(pnl, 2), 
            'qty': qty, 
            'type': exit_type
        })

# --- PARÁMETROS OPTIMIZADOS ---
params = {
    'adx_min': 23,           
    'min_atr': 2.5, 
    'max_atr': 15.0, 
    'qc_threshold': 0.8,     
    'risk_usd': 2000, 
    'max_contracts': 8, 
    'reward_ratio': 1.0, 
    'be_trigger': 1.5, 
    'be_offset': 1.0,
    'max_bars_open': 30      
}

tester = MicrostructureBacktester(FILE_PATH)
trades = tester.run(START_DATE, END_DATE, params)

if not trades.empty:
    trades['pnl_acum'] = trades['pnl'].cumsum()
    print(f"\n✅ BACKTEST FINALIZADO (DATASET V2):")
    print(f"{'='*65}")
    print(f"Total Trades: {len(trades)}")
    print(f"Win Rate:     {(trades['pnl'] > 0).mean():.2%}")
    print(f"PnL Neto:     ${trades['pnl'].sum():,.2f}")
    print(f"{'='*65}\n")
    print(trades[['time', 'points', 'pnl', 'qty', 'type', 'pnl_acum']])
else:
    print("\n❌ No se detectaron trades con los parámetros actuales.")
```

```python
import google.generativeai as genai
from IPython.core.magic import register_line_cell_magic
import time

genai.configure(api_key="AIzaSyB-2w2ImSC3HD25go4zDp717A1-87-VwVo")

# 2.0 Flash es el balance perfecto entre estabilidad y cuota gratuita
modelo_estable = genai.GenerativeModel('gemini-2.0-flash')

@register_line_cell_magic
def gemini(line, cell=None):
    prompt = cell if cell else line
    intentos = 2
    for i in range(intentos):
        try:
            response = modelo_estable.generate_content(prompt)
            # Verificación de seguridad de la respuesta
            if response.candidates and response.candidates[0].content.parts:
                print(response.candidates[0].content.parts[0].text)
                break
            else:
                print("La IA devolvió una respuesta vacía o bloqueada por seguridad.")
        except Exception as e:
            if "429" in str(e):
                print(f"Cuota excedida. Esperando 5 segundos... (Intento {i+1}/{intentos})")
                time.sleep(5)
            else:
                print(f"Error técnico: {e}")
                break

print("✅ Magia %%gemini configurada con Gemini 2.0 Flash (Más estable).")
```

```python
import google.generativeai as genai
from IPython.core.magic import register_line_cell_magic
import time

# Usamos una lista de modelos por orden de preferencia
MODELOS_ORDENADOS = [
    'gemini-1.5-flash-8b', # El más ligero y con cuota más alta
    'gemini-1.5-flash',    # El estándar estable
    'gemini-2.0-flash-lite' # Nueva versión eficiente
]

@register_line_cell_magic
def gemini(line, cell=None):
    prompt = cell if cell else line
    exito = False
    
    for m_id in MODELOS_ORDENADOS:
        if exito: break
        try:
            model = genai.GenerativeModel(m_id)
            response = model.generate_content(prompt)
            if response.text:
                print(f"--- Respuesta via {m_id} ---")
                print(response.text)
                exito = True
        except Exception as e:
            if "429" in str(e):
                continue # Pasa al siguiente modelo si no hay cuota
            else:
                print(f"Error con {m_id}: {e}")
    
    if not exito:
        print("❌ Todos los modelos agotaron su cuota. Te recomiendo esperar 60 segundos o cambiar de API Key.")

print("🚀 Magia configurada con Auto-Fallback (Salto automático de modelo).")
```
